```python
qas = """
***
கணினியின் தேவை தற்போதைய காலங்களில் மிகவும் அதிகரித்துள்ளது. அரசு அலுவலகம் தனியார் அலுவலகம் வங்கிகள் கல்லூரிகள் பள்ளிகள் என கணினி இல்லாத இடமே தற்போது இல்லை. மனித சமூகம் கணினியை நம்பி தன் வாழ்க்கையை ஒப்படைத்து பல வருடங்கள் ஆகிவிட்டன. மனிதனுக்கு உதவும் வகையில் வேளாண்மை, தொழில்நுட்பவியல், பொறியியல், மருத்துவம், ராணுவம், வானியல் ஆராய்ச்சி என அனைத்து இடங்களிலும் கணினி தனது சேவையை செய்கிறது.
##
எதனுடைய தேவை தற்போதைய காலங்களில் மிகவும் அதிகரித்துள்ளது?
##
கணினியின் தேவை
***
கணினியின் தேவை தற்போதைய காலங்களில் மிகவும் அதிகரித்துள்ளது. அரசு அலுவலகம் தனியார் அலுவலகம் வங்கிகள் கல்லூரிகள் பள்ளிகள் என கணினி இல்லாத இடமே தற்போது இல்லை. மனித சமூகம் கணினியை நம்பி தன் வாழ்க்கையை ஒப்படைத்து பல வருடங்கள் ஆகிவிட்டன. மனிதனுக்கு உதவும் வகையில் வேளாண்மை, தொழில்நுட்பவியல், பொறியியல், மருத்துவம், ராணுவம், வானியல் ஆராய்ச்சி என அனைத்து இடங்களிலும் கணினி தனது சேவையை செய்கிறது.
##
எது இல்லாத இடமே தற்போது இல்லை?
##
கணினி
***
கணினியின் தேவை தற்போதைய காலங்களில் மிகவும் அதிகரித்துள்ளது. அரசு அலுவலகம் தனியார் அலுவலகம் வங்கிகள் கல்லூரிகள் பள்ளிகள் என கணினி இல்லாத இடமே தற்போது இல்லை. மனித சமூகம் கணினியை நம்பி தன் வாழ்க்கையை ஒப்படைத்து பல வருடங்கள் ஆகிவிட்டன. மனிதனுக்கு உதவும் வகையில் வேளாண்மை, தொழில்நுட்பவியல், பொறியியல், மருத்துவம், ராணுவம், வானியல் ஆராய்ச்சி என அனைத்து இடங்களிலும் கணினி தனது சேவையை செய்கிறது.
##
 மனித சமூகம் எதை நம்பி தன் வாழ்க்கையை ஒப்படைத்து பல வருடங்கள் ஆகிவிட்டன?
##
கணினியை நம்பி
***
கணினியின் தேவை தற்போதைய காலங்களில் மிகவும் அதிகரித்துள்ளது. அரசு அலுவலகம் தனியார் அலுவலகம் வங்கிகள் கல்லூரிகள் பள்ளிகள் என கணினி இல்லாத இடமே தற்போது இல்லை. மனித சமூகம் கணினியை நம்பி தன் வாழ்க்கையை ஒப்படைத்து பல வருடங்கள் ஆகிவிட்டன. மனிதனுக்கு உதவும் வகையில் வேளாண்மை, தொழில்நுட்பவியல், பொறியியல், மருத்துவம், ராணுவம், வானியல் ஆராய்ச்சி என அனைத்து இடங்களிலும் கணினி தனது சேவையை செய்கிறது.
##
கணினி தனது சேவையை எதில் செய்கிறது?
##
வேளாண்மை, தொழில்நுட்பவியல், பொறியியல், மருத்துவம், ராணுவம், வானியல் ஆராய்ச்சி
***
மனித சமூகம் தகவல்களை சேமிக்க தொடங்கிய நாள் முதலாக கணினியின் ஆதிக்கம் தொடங்கியது. சிறிய அளவில் தொடங்கிய கணினியின் வளர்ச்சி தற்போது விஸ்வரூபம் எடுத்துள்ளது. 20 வருடங்களுக்கு முன்னதாக இருந்த கணினியை விட தற்போது ஆயிரம் மடங்கு சக்தி வாய்ந்த கணினி நமது கைகளில் தவழ்கிறது.
##
எத்தனை வருடங்களுக்கு முன்னதாக இருந்த கணினியை விட தற்போது ஆயிரம் மடங்கு சக்தி வாய்ந்த கணினி நமது கைகளில் தவழ்கிறது?
##
20
***
வேளாண்மை துறையில் சென்ற ஆண்டின் வளர்ச்சி புதிய வேளாண் முயற்சிகள் போன்றவற்றை இணையத்தில் பகிரவும் முந்தைய காலங்களில் ஏற்பட்ட வேளாண் வளர்ச்சி மற்றும் தாக்கங்களை அடுத்த தலைமுறைக்கு எடுத்துச் செல்லவும் கணினி பயன்படுத்தப்படுகிறது அரசு.தற்சமயம் கணினி வழியாக வேளாண் பாடங்களை விவசாயிகளுக்கு உரிய நேரத்தில் கற்பிக்கிறது. கணினியின் உதவி கொண்டு வேளாண்மை வளர்ச்சி என்பது தற்போது தேசத்திற்கு மிகவும் தேவையான ஒன்றாகும்.
##
எது வழியாக வேளாண் பாடங்களை விவசாயிகளுக்கு உரிய நேரத்தில் கற்பிக்கிறது?
##
கணினி வழியாக
***
தற்சமயம் கணினி வழியாக வேளாண் பாடங்களை விவசாயிகளுக்கு உரிய நேரத்தில் கற்பிக்கிறது. கணினியின் உதவி கொண்டு வேளாண்மை வளர்ச்சி என்பது தற்போது தேசத்திற்கு மிகவும் தேவையான ஒன்றாகும்.
##
எதன் உதவி கொண்டு வேளாண்மை வளர்ச்சி என்பது தற்போது தேசத்திற்கு மிகவும் தேவையான ஒன்றாகும்?
##
கணினியின் உதவி
***
மருத்துவத்துறையில் கணினியின் பங்கு மேலும் மேலும் அதிகரித்து கொண்டே உள்ளது எக்ஸ்ரே ஸ்கேன் போன்ற மருத்துவ வேலைகள் தற்போது கணினியின் உதவியுடன் மிக வேகமாக நடைபெறுகிறது. மருத்துவருக்கு உதவுவதோடு மட்டுமல்லாமல் மருந்து உட்கொள்ளும் அவருக்கும் தன்னம்பிக்கையை ஏற்படுத்தும் வகையில் மருந்துகள் பற்றிய ஆராய்ச்சி மற்றும் மருந்துகளை பற்றிய விவரங்களை கணினி மூலமாக நாம் தெரிந்து கொள்ளலாம். புத்தம் புதிய வியாதிகள் வரும் பொழுது அவற்றை எதிர்கொள்ளும் திறனை கணினி மற்றும் இணையம் மூலமாக மக்களுக்கு தெரியப்படுத்துகிறது அரசு. 2019 ஆம் ஆண்டு ஏற்பட்ட கோரோனோ பெரும் தொற்று பற்றிய ஆய்வுகளை உலக அளவில் தெரிய செய்தது இணையமே. இணையம் மூலமாகவே நோய்களை கட்டுப்படுத்தும் முயற்சிகளில் அரசு அதிகம் செலவிட்டது குறிப்பிடத்தக்கது
##
எந்த கணினியின் பங்கு மேலும் மேலும் அதிகரித்து கொண்டே உள்ளது?
##
மருத்துவத்துறையில்
***
மருத்துவத்துறையில் கணினியின் பங்கு மேலும் மேலும் அதிகரித்து கொண்டே உள்ளது எக்ஸ்ரே ஸ்கேன் போன்ற மருத்துவ வேலைகள் தற்போது கணினியின் உதவியுடன் மிக வேகமாக நடைபெறுகிறது. மருத்துவருக்கு உதவுவதோடு மட்டுமல்லாமல் மருந்து உட்கொள்ளும் அவருக்கும் தன்னம்பிக்கையை ஏற்படுத்தும் வகையில் மருந்துகள் பற்றிய ஆராய்ச்சி மற்றும் மருந்துகளை பற்றிய விவரங்களை கணினி மூலமாக நாம் தெரிந்து கொள்ளலாம். புத்தம் புதிய வியாதிகள் வரும் பொழுது அவற்றை எதிர்கொள்ளும் திறனை கணினி மற்றும் இணையம் மூலமாக மக்களுக்கு தெரியப்படுத்துகிறது அரசு. 2019 ஆம் ஆண்டு ஏற்பட்ட கோரோனோ பெரும் தொற்று பற்றிய ஆய்வுகளை உலக அளவில் தெரிய செய்தது இணையமே. இணையம் மூலமாகவே நோய்களை கட்டுப்படுத்தும் முயற்சிகளில் அரசு அதிகம் செலவிட்டது 
##
எது போன்ற மருத்துவ வேலைகள் தற்போது கணினியின் உதவியுடன் மிக வேகமாக நடைபெறுகிறது?
##
ஸ்கேன் போன்ற
***
மருத்துவத்துறையில் கணினியின் பங்கு மேலும் மேலும் அதிகரித்து கொண்டே உள்ளது எக்ஸ்ரே ஸ்கேன் போன்ற மருத்துவ வேலைகள் தற்போது கணினியின் உதவியுடன் மிக வேகமாக நடைபெறுகிறது. மருத்துவருக்கு உதவுவதோடு மட்டுமல்லாமல் மருந்து உட்கொள்ளும் அவருக்கும் தன்னம்பிக்கையை ஏற்படுத்தும் வகையில் மருந்துகள் பற்றிய ஆராய்ச்சி மற்றும் மருந்துகளை பற்றிய விவரங்களை கணினி மூலமாக நாம் தெரிந்து கொள்ளலாம். புத்தம் புதிய வியாதிகள் வரும் பொழுது அவற்றை எதிர்கொள்ளும் திறனை கணினி மற்றும் இணையம் மூலமாக மக்களுக்கு தெரியப்படுத்துகிறது அரசு. 2019 ஆம் ஆண்டு ஏற்பட்ட கோரோனோ பெரும் தொற்று பற்றிய ஆய்வுகளை உலக அளவில் தெரிய செய்தது இணையமே. இணையம் மூலமாகவே நோய்களை கட்டுப்படுத்தும் முயற்சிகளில் அரசு அதிகம் செலவிட்டது 
##
 புத்தம் புதிய வியாதிகள் வரும் பொழுது அவற்றை எதிர்கொள்ளும் திறனை கணினி மற்றும் எதன் மூலமாக மக்களுக்கு தெரியப்படுத்துகிறது?
##
இணையம் மூலமாக
***
2019 ஆம் ஆண்டு ஏற்பட்ட கோரோனோ பெரும் தொற்று பற்றிய ஆய்வுகளை உலக அளவில் தெரிய செய்தது இணையமே. இணையம் மூலமாகவே நோய்களை கட்டுப்படுத்தும் முயற்சிகளில் அரசு அதிகம் செலவிட்டது 
##
எந்த ஆண்டு ஏற்பட்ட கோரோனோ பெரும் தொற்று பற்றிய ஆய்வுகளை உலக அளவில் தெரிய செய்தது?
##
2019
***
2019 ஆம் ஆண்டு ஏற்பட்ட கோரோனோ பெரும் தொற்று பற்றிய ஆய்வுகளை உலக அளவில் தெரிய செய்தது இணையமே. இணையம் மூலமாகவே நோய்களை கட்டுப்படுத்தும் முயற்சிகளில் அரசு அதிகம் செலவிட்டது 
##
எது மூலமாக நோய்களை கட்டுப்படுத்தும் முயற்சிகளில் அரசு அதிகம் செலவிட்டது?
##
இணையம் மூலமாக
***
இராணுவம் மற்றும் பாதுகாப்பு துறையில் கணினியின் பங்கு இன்றியமையாததாக மாறிவிட்டது . இன்றைய காலகட்டங்களில் ராணுவத் தளவாடங்கள் அனைத்தும் கணினியின் புதிய ஏற்பாட்டில் இயங்குகின்றன . துல்லியமாக தாக்கும் காலங்களையும் அதிவேக ஆயுதங்களையும் கணினியின் உதவி கொண்டே இயக்கமுடியும்.
##
எந்த துறையில் கணினியின் பங்கு இன்றியமையாததாக மாறிவிட்டது?
##
இராணுவம் மற்றும் பாதுகாப்பு
***
இராணுவம் மற்றும் பாதுகாப்பு துறையில் கணினியின் பங்கு இன்றியமையாததாக மாறிவிட்டது . இன்றைய காலகட்டங்களில் ராணுவத் தளவாடங்கள் அனைத்தும் கணினியின் புதிய ஏற்பாட்டில் இயங்குகின்றன . துல்லியமாக தாக்கும் காலங்களையும் அதிவேக ஆயுதங்களையும் கணினியின் உதவி கொண்டே இயக்கமுடியும்.
##
இன்றைய காலகட்டங்களில் ராணுவத் தளவாடங்கள் அனைத்தும் எதன் புதிய ஏற்பாட்டில் இயங்குகின்றன?
##
கணினியின் புதிய ஏற்பாட்டில்
***
இராணுவம் மற்றும் பாதுகாப்பு துறையில் கணினியின் பங்கு இன்றியமையாததாக மாறிவிட்டது . இன்றைய காலகட்டங்களில் ராணுவத் தளவாடங்கள் அனைத்தும் கணினியின் புதிய ஏற்பாட்டில் இயங்குகின்றன . துல்லியமாக தாக்கும் காலங்களையும் அதிவேக ஆயுதங்களையும் கணினியின் உதவி கொண்டே இயக்கமுடியும்.
##
கணினியின் உதவி கொண்டே எதை இயக்கமுடியும்?
##
துல்லியமாக தாக்கும் காலங்களையும் அதிவேக ஆயுதங்களையும்
***
கணினி இல்லாமல் ஒரு தனி மனிதன் வாழ்வது கடினம் என்ற அளவுக்கு உயர்ந்த கணினியால் நமக்கு ஏற்படும் அபாயம் இருக்கத்தான் செய்கிறது. அதிக நேரம் கணினியை உபயோகித்தல் தனிநபர் விஷயங்களை திருடுதல் போன்ற இணைய வழி குற்றங்கள் நடக்கவே செய்கிறது.
##
கணினி இல்லாமல் யார் வாழ்வது கடினம்?
##
ஒரு தனி மனிதன் 
***
கணினி இல்லாமல் ஒரு தனி மனிதன் வாழ்வது கடினம் என்ற அளவுக்கு உயர்ந்த கணினியால் நமக்கு ஏற்படும் அபாயம் இருக்கத்தான் செய்கிறது. அதிக நேரம் கணினியை உபயோகித்தல் தனிநபர் விஷயங்களை திருடுதல் போன்ற இணைய வழி குற்றங்கள் நடக்கவே செய்கிறது. கணினி அறிவு பெற்ற ஒரு நபர் மூலம் இத்தகைய திருட்டுக்களை தடுக்க முடியும் .வங்கிக் கணக்குகளில் இருந்து திருடப்படுவது சாதாரண ஒரு நிகழ்வாக தற்சமயம் உள்ளது. 
##
அதிக நேரம் எதை உபயோகித்தல் தனிநபர் விஷயங்களை திருடுதல் போன்ற இணைய வழி குற்றங்கள் நடக்கவே செய்கிறது
##
கணினியை
***
கணினி இல்லாமல் ஒரு தனி மனிதன் வாழ்வது கடினம் என்ற அளவுக்கு உயர்ந்த கணினியால் நமக்கு ஏற்படும் அபாயம் இருக்கத்தான் செய்கிறது. அதிக நேரம் கணினியை உபயோகித்தல் தனிநபர் விஷயங்களை திருடுதல் போன்ற இணைய வழி குற்றங்கள் நடக்கவே செய்கிறது. கணினி அறிவு பெற்ற ஒரு நபர் மூலம் இத்தகைய திருட்டுக்களை தடுக்க முடியும் .வங்கிக் கணக்குகளில் இருந்து திருடப்படுவது சாதாரண ஒரு நிகழ்வாக தற்சமயம் உள்ளது. இவற்றை கணினி மூலமாக ஆய்வுசெய்து தடுக்கமுடியும் இணையவழி குற்றங்களை தடுக்க ஒவ்வொரு அரசும் அதிகமான பொருட்செலவு செய்கிறது 
கணினி ஒரு மின்னனு சாதனம் என்ற நிலைமாறி மனிதனுக்கு உதவும் இன்னொரு விலங்கு என்ற அளவுக்கு உயர்ந்துள்ளது.
##
யார் பெற்ற ஒரு நபர் மூலம் இத்தகைய திருட்டுக்களை தடுக்க முடியும்?
##
கணினி அறிவு 
***
கணினி இல்லாமல் ஒரு தனி மனிதன் வாழ்வது கடினம் என்ற அளவுக்கு உயர்ந்த கணினியால் நமக்கு ஏற்படும் அபாயம் இருக்கத்தான் செய்கிறது. அதிக நேரம் கணினியை உபயோகித்தல் தனிநபர் விஷயங்களை திருடுதல் போன்ற இணைய வழி குற்றங்கள் நடக்கவே செய்கிறது. கணினி அறிவு பெற்ற ஒரு நபர் மூலம் இத்தகைய திருட்டுக்களை தடுக்க முடியும் .வங்கிக் கணக்குகளில் இருந்து திருடப்படுவது சாதாரண ஒரு நிகழ்வாக தற்சமயம் உள்ளது. இவற்றை கணினி மூலமாக ஆய்வுசெய்து தடுக்கமுடியும் இணையவழி குற்றங்களை தடுக்க ஒவ்வொரு அரசும் அதிகமான பொருட்செலவு செய்கிறது 
கணினி ஒரு மின்னனு சாதனம் என்ற நிலைமாறி மனிதனுக்கு உதவும் இன்னொரு விலங்கு என்ற அளவுக்கு உயர்ந்துள்ளது. மிக சக்திவாய்ந்த குழப்பமான கணக்கு மற்றும் மனித வேலைகளை சுலபமாக முடிப்பதற்கு கணினி நமக்கு உதவுகிறது மனித கலாச்சாரத்தில் கணினி கண்டுபிடிப்பு அசுர வேக அறிவியல் 
##
எதில் இருந்து திருடப்படுவது சாதாரண ஒரு நிகழ்வாக தற்சமயம் உள்ளது?
##
வங்கிக் கணக்குகளில் இருந்து
***
கணினி அறிவு பெற்ற ஒரு நபர் மூலம் இத்தகைய திருட்டுக்களை தடுக்க முடியும் .வங்கிக் கணக்குகளில் இருந்து திருடப்படுவது சாதாரண ஒரு நிகழ்வாக தற்சமயம் உள்ளது. இவற்றை கணினி மூலமாக ஆய்வுசெய்து தடுக்கமுடியும் இணையவழி குற்றங்களை தடுக்க ஒவ்வொரு அரசும் அதிகமான பொருட்செலவு செய்கிறது 
கணினி ஒரு மின்னனு சாதனம் என்ற நிலைமாறி மனிதனுக்கு உதவும் இன்னொரு விலங்கு என்ற அளவுக்கு உயர்ந்துள்ளது. மிக சக்திவாய்ந்த குழப்பமான கணக்கு மற்றும் மனித வேலைகளை சுலபமாக முடிப்பதற்கு கணினி நமக்கு உதவுகிறது மனித கலாச்சாரத்தில் கணினி கண்டுபிடிப்பு அசுர வேக அறிவியல் 
##
 இவற்றை கணினி மூலமாக ஆய்வுசெய்து தடுக்கமுடியும் இணையவழி குற்றங்களை தடுக்க ஒவ்வொரு அரசும் அதிகமான என்ன செலவு செய்கிறது?
##
பொருட்செலவு செய்கிறது
***
கணினி இல்லாமல் ஒரு தனி மனிதன் வாழ்வது கடினம் என்ற அளவுக்கு உயர்ந்த கணினியால் நமக்கு ஏற்படும் அபாயம் இருக்கத்தான் செய்கிறது. அதிக நேரம் கணினியை உபயோகித்தல் தனிநபர் விஷயங்களை திருடுதல் போன்ற இணைய வழி குற்றங்கள் நடக்கவே செய்கிறது. கணினி அறிவு பெற்ற ஒரு நபர் மூலம் இத்தகைய திருட்டுக்களை தடுக்க முடியும் .வங்கிக் கணக்குகளில் இருந்து திருடப்படுவது சாதாரண ஒரு நிகழ்வாக தற்சமயம் உள்ளது. இவற்றை கணினி மூலமாக ஆய்வுசெய்து தடுக்கமுடியும் இணையவழி குற்றங்களை தடுக்க ஒவ்வொரு அரசும் அதிகமான பொருட்செலவு செய்கிறது 
கணினி ஒரு மின்னனு சாதனம் என்ற நிலைமாறி மனிதனுக்கு உதவும் இன்னொரு விலங்கு என்ற அளவுக்கு உயர்ந்துள்ளது. மிக சக்திவாய்ந்த குழப்பமான கணக்கு மற்றும் மனித வேலைகளை சுலபமாக முடிப்பதற்கு கணினி நமக்கு உதவுகிறது மனித கலாச்சாரத்தில் கணினி கண்டுபிடிப்பு அசுர வேக அறிவியல் 
##
கணினி ஒரு என்ன சாதனம்?
##
மின்னனு சாதனம்
***
 இவற்றை கணினி மூலமாக ஆய்வுசெய்து தடுக்கமுடியும் இணையவழி குற்றங்களை தடுக்க ஒவ்வொரு அரசும் அதிகமான பொருட்செலவு செய்கிறது 
கணினி ஒரு மின்னனு சாதனம் என்ற நிலைமாறி மனிதனுக்கு உதவும் இன்னொரு விலங்கு என்ற அளவுக்கு உயர்ந்துள்ளது. மிக சக்திவாய்ந்த குழப்பமான கணக்கு மற்றும் மனித வேலைகளை சுலபமாக முடிப்பதற்கு கணினி நமக்கு உதவுகிறது மனித கலாச்சாரத்தில் கணினி கண்டுபிடிப்பு அசுர வேக அறிவியல் 
##
மிக சக்திவாய்ந்த குழப்பமான கணக்கு மற்றும் யார் வேலைகளை சுலபமாக முடிப்பதற்கு?
##
 மனித வேலைகளை
***
பள்ளி கல்லூரிகளில் கணினியின் பங்கு எப்போதும் குறைந்ததில்லை . பள்ளிக்குச் சென்றுதான் படிக்கவேண்டும் என்ற நிலைமை மாறி இணைய வழியில் பாடம் கற்கும் நாளைக்கு நாம் தள்ளப்பட்டு விட்டோம். கோரோனோ தொற்று பரவும் காலங்களில் இளம் வயதினரை பள்ளி கல்லூரிக்கு அனுப்பி தொற்று ஏற்படாமல் இருக்கும் வகையில் அரசு இணைய வழி மூலமாக பாடங்களை நடத்தியது குறிப்பிடத்தக்கது. அமைந்துவிட்டது
##
எங்கு சென்றுதான் படிக்கவேண்டும் என்ற நிலைமை மாறி இணைய வழியில் பாடம் கற்கும் நாளைக்கு நாம் தள்ளப்பட்டு விட்டோம்
##
பள்ளிக்குச் சென்றுதான்
***
பள்ளி கல்லூரிகளில் கணினியின் பங்கு எப்போதும் குறைந்ததில்லை . பள்ளிக்குச் சென்றுதான் படிக்கவேண்டும் என்ற நிலைமை மாறி இணைய வழியில் பாடம் கற்கும் நாளைக்கு நாம் தள்ளப்பட்டு விட்டோம். கோரோனோ தொற்று பரவும் காலங்களில் இளம் வயதினரை பள்ளி கல்லூரிக்கு அனுப்பி தொற்று ஏற்படாமல் இருக்கும் வகையில் அரசு இணைய வழி மூலமாக பாடங்களை நடத்தியது குறிப்பிடத்தக்கது. இணைய வழி பாடங்கள் சாதாரணமாக ஒரு ஆசிரியர் வகுப்பறையில் நடத்துவது போன்ற மட்டுமில்லாமல் புத்தம் புதிய தொழில்நுட்பம் வழியாக மாணவர்களுக்கு பாடங்களை நடத்துகிறது.
##
யார் இணைய வழி மூலமாக பாடங்களை நடத்தியது குறிப்பிடத்தக்கது?
##
அரசு
***
கோரோனோ தொற்று பரவும் காலங்களில் இளம் வயதினரை பள்ளி கல்லூரிக்கு அனுப்பி தொற்று ஏற்படாமல் இருக்கும் வகையில் அரசு இணைய வழி மூலமாக பாடங்களை நடத்தியது குறிப்பிடத்தக்கது. இணைய வழி பாடங்கள் சாதாரணமாக ஒரு ஆசிரியர் வகுப்பறையில் நடத்துவது போன்ற மட்டுமில்லாமல் புத்தம் புதிய தொழில்நுட்பம் வழியாக மாணவர்களுக்கு பாடங்களை நடத்துகிறது. அனிமேஷன் மற்றும் வீடியோக்கள் மூலமாக மாணவர் சுலபமாக கல்வி கற்கும் திறனை வளர்க்கிறது. இதன் காரணமாகவே ஆசிரியராக இருக்கும் ஒருவருக்கு கணினி அறிவும் தேவை என்ற குறைந்தபட்ச தகுதி தாமாக அமைந்துவிட்டது
##
இணைய வழி பாடங்கள் சாதாரணமாக ஒரு ஆசிரியர் வகுப்பறையில் நடத்துவது போன்ற மட்டுமில்லாமல் எது வழியாக மாணவர்களுக்கு பாடங்களை நடத்துகிறது?
##
புத்தம் புதிய தொழில்நுட்பம் வழியாக
***
கோரோனோ தொற்று பரவும் காலங்களில் இளம் வயதினரை பள்ளி கல்லூரிக்கு அனுப்பி தொற்று ஏற்படாமல் இருக்கும் வகையில் அரசு இணைய வழி மூலமாக பாடங்களை நடத்தியது குறிப்பிடத்தக்கது. இணைய வழி பாடங்கள் சாதாரணமாக ஒரு ஆசிரியர் வகுப்பறையில் நடத்துவது போன்ற மட்டுமில்லாமல் புத்தம் புதிய தொழில்நுட்பம் வழியாக மாணவர்களுக்கு பாடங்களை நடத்துகிறது. அனிமேஷன் மற்றும் வீடியோக்கள் மூலமாக மாணவர் சுலபமாக கல்வி கற்கும் திறனை வளர்க்கிறது. இதன் காரணமாகவே ஆசிரியராக இருக்கும் ஒருவருக்கு கணினி அறிவும் தேவை என்ற குறைந்தபட்ச தகுதி தாமாக அமைந்துவிட்டது
##
எது மூலமாக மாணவர் சுலபமாக கல்வி கற்கும் திறனை வளர்க்கிறது?
##
அனிமேஷன் மற்றும் வீடியோக்கள் மூலமாக
***
கணினி விலை ஒவ்வொரு ஆண்டும் குறைந்து கொண்டே வருகிறது மறுபக்கம் கணினியின் சக்தி ஒவ்வொரு ஆண்டும் உயர்ந்து கொண்டே வருகிறது 20 ஆண்டுகளுக்கு முன்பாக ஒரு லட்ச ரூபாய்க்கு விற்கப்பட்ட கணினியின்  சக்தியைவிட பத்தாயிரம் ரூபாய்க்கு விற்கப்படும் செல்பேசியின் சக்தி இன்று நூறு மடங்கு அதிகமாக உள்ளது
##
எதன் விலை ஒவ்வொரு ஆண்டும் குறைந்து கொண்டே வருகிறது?
##
கணினி விலை
***
கணினி விலை ஒவ்வொரு ஆண்டும் குறைந்து கொண்டே வருகிறது மறுபக்கம் கணினியின் சக்தி ஒவ்வொரு ஆண்டும் உயர்ந்து கொண்டே வருகிறது 20 ஆண்டுகளுக்கு முன்பாக ஒரு லட்ச ரூபாய்க்கு விற்கப்பட்ட கணினியின்  சக்தியைவிட பத்தாயிரம் ரூபாய்க்கு விற்கப்படும் செல்பேசியின் சக்தி இன்று நூறு மடங்கு அதிகமாக உள்ளது
##
எத்தனை ஆண்டுகளுக்கு முன்பாக ஒரு லட்ச ரூபாய்க்கு விற்கப்பட்ட கணினியின்  சக்தியைவிட பத்தாயிரம் ரூபாய்க்கு விற்கப்படும் செல்பேசியின் சக்தி இன்று நூறு மடங்கு அதிகமாக உள்ளது?
##
20
***
கணினி விலை ஒவ்வொரு ஆண்டும் குறைந்து கொண்டே வருகிறது மறுபக்கம் கணினியின் சக்தி ஒவ்வொரு ஆண்டும் உயர்ந்து கொண்டே வருகிறது 20 ஆண்டுகளுக்கு முன்பாக ஒரு லட்ச ரூபாய்க்கு விற்கப்பட்ட கணினியின்  சக்தியைவிட பத்தாயிரம் ரூபாய்க்கு விற்கப்படும் செல்பேசியின் சக்தி இன்று நூறு மடங்கு அதிகமாக உள்ளது
##
20 ஆண்டுகளுக்கு முன்பாக ஒரு லட்ச ரூபாய்க்கு விற்கப்பட்ட கணினியின்  சக்தியைவிட பத்தாயிரம் ரூபாய்க்கு விற்கப்படும் செல்பேசியின் சக்தி இன்று எத்தனை மடங்கு அதிகமாக உள்ளது?
##
நூறு மடங்கு

***
கணினி விலை ஒவ்வொரு ஆண்டும் குறைந்து கொண்டே வருகிறது மறுபக்கம் கணினியின் சக்தி ஒவ்வொரு ஆண்டும் உயர்ந்து கொண்டே வருகிறது 20 ஆண்டுகளுக்கு முன்பாக ஒரு லட்ச ரூபாய்க்கு விற்கப்பட்ட கணினியின்  சக்தியைவிட பத்தாயிரம் ரூபாய்க்கு விற்கப்படும் செல்பேசியின் சக்தி இன்று நூறு மடங்கு அதிகமாக உள்ளது
##
20 ஆண்டுகளுக்கு முன்பாக ஒரு லட்ச ரூபாய்க்கு விற்கப்பட்ட கணினியின்  சக்தியைவிட பத்தாயிரம் ரூபாய்க்கு விற்கப்படும் எந்த சக்தி இன்று நூறு மடங்கு அதிகமாக உள்ளது?
##
செல்பேசியின் சக்தி
***
 கணினியில் சக்திக்கு ஈடு கொடுக்கும் வகையில் இணைய வேகம் தொலைத்தொடர்பு மூலமாக நொடிப்பொழுதில் உலகை சுற்றி வரும் வல்லமை கொண்டதாக மாறி வருகிறது 10 ஆண்டுகளில் இணையத்தின் வேகம் 200 மடங்கு அதிகமாக உள்ளது என்பது குறிப்பிடத்தக்கது இனிவரும் காலங்களில் இது ஆயிரம் மடங்கு வேகமாக செயல்பட உள்ளது
##
எத்தனை ஆண்டுகளில் இணையத்தின் வேகம் 200 மடங்கு அதிகமாக உள்ளது?
##
10 ஆண்டுகளில்
***
 கணினியில் சக்திக்கு ஈடு கொடுக்கும் வகையில் இணைய வேகம் தொலைத்தொடர்பு மூலமாக நொடிப்பொழுதில் உலகை சுற்றி வரும் வல்லமை கொண்டதாக மாறி வருகிறது 10 ஆண்டுகளில் இணையத்தின் வேகம் 200 மடங்கு அதிகமாக உள்ளது என்பது குறிப்பிடத்தக்கது இனிவரும் காலங்களில் இது ஆயிரம் மடங்கு வேகமாக செயல்பட உள்ளது
##
10 ஆண்டுகளில் இணையத்தின் வேகம் எத்தனை மடங்கு அதிகமாக உள்ளது?
##
200 மடங்கு
***
தமிழ், திராவிட மொழிக் குடும்பத்தைச் சேர்ந்த ஒரு மொழியாகும். இக்குடும்பத்தில், இருளா, கைக்காடி, பெட்டக் குறும்பா, சோலகா மற்றும் யெருகுலா என்னும் மொழிகள் அடங்கும். தமிழ் மொழிக் குடும்பம் .
##
தமிழ் எந்த குடும்மத்தை சேர்ந்த ஒரு மொழியாகும் ?
##
திராவிட மொழிக்
***
இந்தியாவின், 11 ஆவது குடியரசு தலைவராக அப்துல் கலாம் பணியாற்றினார். இவர் தமிழ்நாட்டில் உள்ள இராமேஸ்வரம் என்றஇடத்தில் பிறந்து வளர்ந்தார்.
##
இந்தியாவின், 11 ஆவது குடியரசு தலைவராக பணியாற்றியவர்  யார் ?
##
அப்துல் கலாம்
***
இந்தியாவின், 11 ஆவது குடியரசு தலைவராக அப்துல் கலாம் பணியாற்றினார். இவர் தமிழ்நாட்டில் உள்ள இராமேஸ்வரம் என்றஇடத்தில் பிறந்து வளர்ந்தார்.
##
அப்துல் கலாம் எங்கு பிறந்ததார் ?
##
இராமேஸ்வரம் 
***
இரண்டாம் உலகப்போர் என்பது 1939-45 காலகட்டத்தில் நடைபெற்ற ஒரு போர்.ஏறத்தாழ 10 கோடி போர் வீரர்கள் இதில் பங்கு கொண்டனர்.அமெரிக்காவும் இங்கிலாந்தும் சேர்ந்து ஆகஸ்ட் மாத துவக்கத்தில் ஜப்பான் நாட்டின் ஹிரோஷிமா, நாகசாகி போன்ற நகரங்களில் அடுத்தடுத்து அணுக்குண்டு மழைபொழிந்தது.1945ம் ஆண்டு ஆகஸ்ட் மாதம் 15 ம் ஜப்பான் போரை விடுத்து சரணடைய ஒப்புக்கொண்டது.
##
இரண்டாம் உலகப்போர் எந்த  காலகட்டத்தில் நடைபெற்றது ?
##
1939-45 காலகட்டத்தில்
***
இரண்டாம் உலகப்போர் என்பது 1939-45 காலகட்டத்தில் நடைபெற்ற ஒரு போர்.ஏறத்தாழ 10 கோடி போர் வீரர்கள் இதில் பங்கு கொண்டனர்.அமெரிக்காவும் இங்கிலாந்தும் சேர்ந்து ஆகஸ்ட் மாத துவக்கத்தில் ஜப்பான் நாட்டின் ஹிரோஷிமா, நாகசாகி போன்ற நகரங்களில் அடுத்தடுத்து அணுக்குண்டு மழைபொழிந்தது.1945ம் ஆண்டு ஆகஸ்ட் மாதம் 15 ம் ஜப்பான் போரை விடுத்து சரணடைய ஒப்புக்கொண்டது.
##
எத்தனை கோடி போர் வீரர்கள் இதில் பங்கு கொண்டனர் ?
##
ஏறத்தாழ 10 கோடி
***
இரண்டாம் உலகப்போர் என்பது 1939-45 காலகட்டத்தில் நடைபெற்ற ஒரு போர்.ஏறத்தாழ 10 கோடி போர் வீரர்கள் இதில் பங்கு கொண்டனர்.அமெரிக்காவும் இங்கிலாந்தும் சேர்ந்து ஆகஸ்ட் மாத துவக்கத்தில் ஜப்பான் நாட்டின் ஹிரோஷிமா, நாகசாகி போன்ற நகரங்களில் அடுத்தடுத்து அணுக்குண்டு மழைபொழிந்தது.1945ம் ஆண்டு ஆகஸ்ட் மாதம் 15 ம் ஜப்பான் போரை விடுத்து சரணடைய ஒப்புக்கொண்டது.
##
அமெரிக்காவும் இங்கிலாந்தும் சேர்ந்து எந்த மாத துவக்கத்தில் ஜப்பான் நாட்டின் ஹிரோஷிமா, நாகசாகி போன்ற நகரங்களில் அடுத்தடுத்து அணுக்குண்டு மழைபொழிந்ததன?
##
ஆகஸ்ட் மாத துவக்கத்தில் 
***
இரண்டாம் உலகப்போர் என்பது 1939-45 காலகட்டத்தில் நடைபெற்ற ஒரு போர்.ஏறத்தாழ 10 கோடி போர் வீரர்கள் இதில் பங்கு கொண்டனர்.அமெரிக்காவும் இங்கிலாந்தும் சேர்ந்து ஆகஸ்ட் மாத துவக்கத்தில் ஜப்பான் நாட்டின் ஹிரோஷிமா, நாகசாகி போன்ற நகரங்களில் அடுத்தடுத்து அணுக்குண்டு மழைபொழிந்தது.1945ம் ஆண்டு ஆகஸ்ட் மாதம் 15 ம் ஜப்பான் போரை விடுத்து சரணடைய ஒப்புக்கொண்டது.
##
எந்த நாடுகள் ஆகஸ்ட் மாத துவக்கத்தில் ஜப்பான் நாட்டின் ஹிரோஷிமா, நாகசாகி போன்ற நகரங்களில் அடுத்தடுத்து அணுக்குண்டு மழைபொழிந்தன ?
##
அமெரிக்காவும் இங்கிலாந்தும் 
***
இரண்டாம் உலகப்போர் என்பது 1939-45 காலகட்டத்தில் நடைபெற்ற ஒரு போர்.ஏறத்தாழ 10 கோடி போர் வீரர்கள் இதில் பங்கு கொண்டனர்.அமெரிக்காவும் இங்கிலாந்தும் சேர்ந்து ஆகஸ்ட் மாத துவக்கத்தில் ஜப்பான் நாட்டின் ஹிரோஷிமா, நாகசாகி போன்ற நகரங்களில் அடுத்தடுத்து அணுக்குண்டு மழைபொழிந்தது.1945ம் ஆண்டு ஆகஸ்ட் மாதம் 15 ம் ஜப்பான் போரை விடுத்து சரணடைய ஒப்புக்கொண்டது.
##
அமெரிக்காவும் இங்கிலாந்தும் சேர்ந்து ஆகஸ்ட் மாத துவக்கத்தில் ஜப்பான் நாட்டின் எந்த நகரங்களில் அடுத்தடுத்து அணுக்குண்டு மழைபொழிந்தன?
##
ஹிரோஷிமா, நாகசாகி
***
10 அணிகளைக் கொண்ட ஒரு குழுவில் ஒவ்வோர் அணியும் ஏனைய ஒன்பது அணிகளுடன் ஒரு முறை மோதின.தொடர் முழுவதும் சிறப்பாக ஆடிய வீரர்களைக் கொண்டு தொடரின் அணி உருவாக்கப்படுகிறது. அந்த வகையில் 15 சூலை 2019 நாளன்று கேன் வில்லியம்சன் தலைமையிலான தொடரின் அணியை ஐசிசி அறிவித்தது.
##
ஒரு குழுவில் எத்தனை அணிகள் உள்ளன ?
##
10
***
10 அணிகளைக் கொண்ட ஒரு குழுவில் ஒவ்வோர் அணியும் ஏனைய ஒன்பது அணிகளுடன் ஒரு முறை மோதின.தொடர் முழுவதும் சிறப்பாக ஆடிய வீரர்களைக் கொண்டு தொடரின் அணி உருவாக்கப்படுகிறது. அந்த வகையில் 15 சூலை 2019 நாளன்று கேன் வில்லியம்சன் தலைமையிலான தொடரின் அணியை ஐசிசி அறிவித்தது.
##
யார் தலைமையிலான தொடரின் அணியை ஐசிசி அறிவித்தது?
##
கேன் வில்லியம்சன்
***
கோவிட்-19, இது சீனாவில் கண்டறியப்பட்ட கொரோனாவைரசு மனிதர்களிடம் ஏற்படுத்தும் தொற்று நோயாகும். சீனாவின் ஊகான் நகரத்தில் இது ஏற்படுத்திய கொடிய தாக்கத்திற்குப் பின்பு தான் இந்த பெயர் வைக்கப்பட்டது.31 திசம்பர் 2019 அன்று, உலக சுகாதார அமைப்பால் அதிகாரப்பூர்வமாக அறிவிக்கப்பட்ட 2019-nCoV என்ற கொரோனாவைரசின் ஒரு புதிய திரிபு சீனாவின் ஊகான் மாகாணத்தில் கண்டுபிடிக்கப்பட்டது.
##
கோவிட்-19 எங்கு கண்டறியப்பட்ட கொரோனாவைரசு?
##
சீனா
***
இந்தியன் பிரீமியர் லீக் (ஐபிஎல்) என்பது இந்தியாவில் விளையாடப்படும் தொழில்முறை இருபது20 துடுப்பாட்டப் போட்டித் தொடர் ஆகும். இது ஒவ்வொரு ஆண்டும் ஏப்ரல்-மே மாதங்களில் நடைபெறும் .ஐபிஎல் உலகிலேயே அதிக பார்வையாளர்கள் பங்கேற்ற துடுப்பாட்டப் போட்டித் தொடர் ஆகும்.2018ஆம் ஆண்டில் ஐபிஎல்லின் பிராண்ட் மதிப்பு 6.3 பில்லியன் அமெரிக்க டாலராக இருந்தது.ஐபிஎல் தொடரில் இதுவரை மொத்தம் பதினான்கு பருவங்கள் நடைபெற்றுள்ளன.செப்டம்பர் 2018இல் ஐபிஎல் போட்டிகளின் 5 வருட உலகளாவிய ஒளிபரப்பு ஒப்பந்தத்தை ஸ்டார் இந்தியா நிறுவனம் கைப்பற்றியது.
##
ஆண்டின் எந்த மாதங்களில் ஐபிஎல் நடைபெறும்?
##
ஏப்ரல்-மே
***
இந்தியன் பிரீமியர் லீக் (ஐபிஎல்) என்பது இந்தியாவில் விளையாடப்படும் தொழில்முறை இருபது20 துடுப்பாட்டப் போட்டித் தொடர் ஆகும். இது ஒவ்வொரு ஆண்டும் ஏப்ரல்-மே மாதங்களில் நடைபெறும் .ஐபிஎல் உலகிலேயே அதிக பார்வையாளர்கள் பங்கேற்ற துடுப்பாட்டப் போட்டித் தொடர் ஆகும்.2018ஆம் ஆண்டில் ஐபிஎல்லின் பிராண்ட் மதிப்பு 6.3 பில்லியன் அமெரிக்க டாலர் .ஐபிஎல் தொடரில் இதுவரை மொத்தம் பதினான்கு பருவங்கள் நடைபெற்றுள்ளன.செப்டம்பர் 2018இல் ஐபிஎல் போட்டிகளின் 5 வருட உலகளாவிய ஒளிபரப்பு ஒப்பந்தத்தை ஸ்டார் இந்தியா நிறுவனம் கைப்பற்றியது.
##
2018ஆம் ஆண்டில் ஐபிஎல்லின் பிராண்ட் மதிப்பு என்ன ?
##
6.3 பில்லியன் அமெரிக்க டாலர்
***
இந்தியன் பிரீமியர் லீக் (ஐபிஎல்) என்பது இந்தியாவில் விளையாடப்படும் தொழில்முறை இருபது20 துடுப்பாட்டப் போட்டித் தொடர் ஆகும். இது ஒவ்வொரு ஆண்டும் ஏப்ரல்-மே மாதங்களில் நடைபெறும் .ஐபிஎல் உலகிலேயே அதிக பார்வையாளர்கள் பங்கேற்ற துடுப்பாட்டப் போட்டித் தொடர் ஆகும்.2018ஆம் ஆண்டில் ஐபிஎல்லின் பிராண்ட் மதிப்பு 6.3 பில்லியன் அமெரிக்க டாலர் .ஐபிஎல் தொடரில் இதுவரை மொத்தம் பதினான்கு பருவங்கள் நடைபெற்றுள்ளன.செப்டம்பர் 2018இல் ஐபிஎல் போட்டிகளின் 5 வருட உலகளாவிய ஒளிபரப்பு ஒப்பந்தத்தை ஸ்டார் இந்தியா நிறுவனம் கைப்பற்றியது.
##
ஐபிஎல் தொடரில் இதுவரை மொத்தம் எத்தனை பருவங்கள் நடைபெற்றுள்ளன ?
##
பதினான்கு
***
இந்தியன் பிரீமியர் லீக் (ஐபிஎல்) என்பது இந்தியாவில் விளையாடப்படும் தொழில்முறை இருபது20 துடுப்பாட்டப் போட்டித் தொடர் ஆகும். இது ஒவ்வொரு ஆண்டும் ஏப்ரல்-மே மாதங்களில் நடைபெறும் .ஐபிஎல் உலகிலேயே அதிக பார்வையாளர்கள் பங்கேற்ற துடுப்பாட்டப் போட்டித் தொடர் ஆகும்.2018ஆம் ஆண்டில் ஐபிஎல்லின் பிராண்ட் மதிப்பு 6.3 பில்லியன் அமெரிக்க டாலர் .ஐபிஎல் தொடரில் இதுவரை மொத்தம் பதினான்கு பருவங்கள் நடைபெற்றுள்ளன.செப்டம்பர் 2018இல் ஐபிஎல் போட்டிகளின் 5 வருட உலகளாவிய ஒளிபரப்பு ஒப்பந்தத்தை ஸ்டார் இந்தியா நிறுவனம் கைப்பற்றியது.
##
செப்டம்பர் 2018இல் ஐபிஎல் போட்டிகளின் 5 வருட உலகளாவிய ஒளிபரப்பு ஒப்பந்தத்தை எந்த  நிறுவனம் கைப்பற்றியது?
##
ஸ்டார் இந்தியா
***
கோலிவுட் என்று அழைக்கப்படும் தமிழ்த் திரைப்படத்துறை எனப்படுவது தமிழ் நாட்டில் இருந்து இயங்கும் தமிழ் மொழித் திரைப்படத்துறை ஆகும்.முதல் தமிழ் ஊமைத் திரைப்படம் 1918 ஆம் ஆண்டில் இயக்குநர் ஆர். நடராஜ முதலியார் என்பவர் இயக்கிய கீசக வதம் என்ற திரைப்படம் ஆகும்.
##
கோலிவுட் எந்த மொழித் திரைப்படத்துறை?
##
தமிழ் மொழி
***
கோலிவுட் என்று அழைக்கப்படும் தமிழ்த் திரைப்படத்துறை எனப்படுவது தமிழ் நாட்டில் இருந்து இயங்கும் தமிழ் மொழித் திரைப்படத்துறை ஆகும்.முதல் தமிழ் ஊமைத் திரைப்படம் 1918 ஆம் ஆண்டில் இயக்குநர் ஆர். நடராஜ முதலியார் என்பவர் இயக்கிய கீசக வதம் என்ற திரைப்படம் ஆகும்.
##
முதல் தமிழ் ஊமைத் திரைப்படம் எது ?
##
கீசக வதம்
***
சிவ சுப்ரமணிய நாடார் பொறியியல் கல்லூரி (SSN) தமிழ்நாட்டில் உள்ள ஒரு பொறியியல் கல்லூரி. 1996ம் ஆண்டு எச்.சி.எல் கணினி நிறுவனத்தின் நிறுவனர் சிவ நாடாரால் தொடங்கப்பட்டது.இக்கல்லூரி 2011 இல் இந்தியாவின் முதல் 75 பொறியியல் கல்லூரிகளில் 38வது இடத்திலிருப்பதாக அவுட்லுக் இதழ் பிரசுரித்தது.இக்கல்லூரி ராஜீவ் காந்தி சாலையில்  உள்ள காலவாக்கத்தில் அமைக்கப்பட்டுள்ளது.
##
எந்த கல்லூரி ராஜீவ் காந்தி சாலையில்  உள்ள காலவாக்கத்தில் அமைக்கப்பட்டுள்ளது?
##
சிவ சுப்ரமணிய நாடார் பொறியியல் கல்லூரி
***
சிவ சுப்ரமணிய நாடார் பொறியியல் கல்லூரி (SSN) தமிழ்நாட்டில் உள்ள ஒரு பொறியியல் கல்லூரி. 1996ம் ஆண்டு எச்.சி.எல் கணினி நிறுவனத்தின் நிறுவனர் சிவ நாடாரால் தொடங்கப்பட்டது.இக்கல்லூரி 2011 இல் இந்தியாவின் முதல் 75 பொறியியல் கல்லூரிகளில் 38வது இடத்திலிருப்பதாக அவுட்லுக் இதழ் பிரசுரித்தது.இக்கல்லூரி ராஜீவ் காந்தி சாலையில்  உள்ள காலவாக்கத்தில் அமைக்கப்பட்டுள்ளது.
##
சிவ சுப்ரமணிய நாடார் பொறியியல் கல்லூரி எந்த ஆண்டு தொடங்கப்பட்டது?
##
1996
***
சிவ சுப்ரமணிய நாடார் பொறியியல் கல்லூரி (SSN) தமிழ்நாட்டில் உள்ள ஒரு பொறியியல் கல்லூரி. 1996ம் ஆண்டு எச்.சி.எல் கணினி நிறுவனத்தின் நிறுவனர் சிவ நாடாரால் தொடங்கப்பட்டது.இக்கல்லூரி 2011 இல் இந்தியாவின் முதல் 75 பொறியியல் கல்லூரிகளில் 38வது இடத்திலிருப்பதாக அவுட்லுக் இதழ் பிரசுரித்தது.இக்கல்லூரி ராஜீவ் காந்தி சாலையில்  உள்ள காலவாக்கத்தில் அமைக்கப்பட்டுள்ளது.
##
சிவ சுப்ரமணிய நாடார் பொறியியல் கல்லூரி யாரால்  தொடங்கப்பட்டது?
##
சிவ நாடாரால்
***
தமிழ் தமிழர்களினதும் தமிழ் பேசும் பலரின் தாய்மொழி ஆகும். தமிழ், உலகின் உள்ள முதன்மையான மொழிகளில் ஒன்றும் செம்மொழியும் ஆகும். இந்தியா, இலங்கை, மலேசியா, சிங்கப்பூர் ஆகிய நாடுகளில் அதிக அளவிலும், ஐக்கிய அரபு அமீரகம், தென்னாப்பிரிக்கா, மொரிசியசு, பிசி, இரீயூனியன், திரினிடாடு போன்ற நாடுகளில் சிறிய அளவிலும் தமிழ் பேசப்படுகிறது. 1997-ஆம் ஆண்டுப் புள்ளி விவரப்படி உலகம் முழுவதிலும் 8 கோடி (80 மில்லியன்) மக்களால் பேசப்படும் தமிழ், ஒரு மொழியைத் தாய்மொழியாகக் கொண்டு பேசும் மக்களின் எண்ணிக்கை அடிப்படையில் பதினெட்டாவது இடத்தில் உள்ளது. இணையத்தில் அதிகம் பயன்படுத்தப்படும் இந்திய மொழிகளில் தமிழ் முதன்மையாக உள்ளதாக 2017-ஆம் ஆண்டு நடைபெற்ற கூகுள் கணக்கெடுப்பில் தெரிய வந்தது.
இரண்டாயிரத்து ஐந்நூறு ஆண்டுகளுக்கும் மேல் பழமை வாய்ந்த இலக்கிய மரபைக் கொண்டுள்ள தமிழ் மொழி, தற்போது வழக்கில் இருக்கும் ஒரு சில செம்மொழிகளில் இதுவும் ஒன்றாகும். மேலும் கவனமாகப் பழைய அமைப்புக்களைக் காக்கும் மரபினாலும் பழங்கால இலக்கிய நடை கூட மக்களால் புரிந்து கொள்ளும் நிலையில் உள்ளது. எடுத்துக்காட்டாக, பள்ளிக் குழந்தைகள் சிறுவயதில் கற்கும் அகர வரிசையான ஆத்திசூடி 1,000 ஆண்டுகளுக்கு முன்னர் இயற்றப்பட்டது. திருக்குறள் ஏறத்தாழ 2,000 ஆண்டுகளுக்கு முன் இயற்றப்பட்டது.
##
ஒரு மொழியைத் தாய்மொழியாகக் கொண்டு பேசும் மக்களின் எண்ணிக்கை அடிப்படையில் எத்தனையாவது  இடத்தில் உள்ளது ?
##
பதினெட்டாவது
***
தமிழ் தமிழர்களினதும் தமிழ் பேசும் பலரின் தாய்மொழி ஆகும். தமிழ், உலகின் உள்ள முதன்மையான மொழிகளில் ஒன்றும் செம்மொழியும் ஆகும். இந்தியா, இலங்கை, மலேசியா, சிங்கப்பூர் ஆகிய நாடுகளில் அதிக அளவிலும், ஐக்கிய அரபு அமீரகம், தென்னாப்பிரிக்கா, மொரிசியசு, பிசி, இரீயூனியன், திரினிடாடு போன்ற நாடுகளில் சிறிய அளவிலும் தமிழ் பேசப்படுகிறது. 1997-ஆம் ஆண்டுப் புள்ளி விவரப்படி உலகம் முழுவதிலும் 8 கோடி (80 மில்லியன்) மக்களால் பேசப்படும் தமிழ், ஒரு மொழியைத் தாய்மொழியாகக் கொண்டு பேசும் மக்களின் எண்ணிக்கை அடிப்படையில் பதினெட்டாவது இடத்தில் உள்ளது. இணையத்தில் அதிகம் பயன்படுத்தப்படும் இந்திய மொழிகளில் தமிழ் முதன்மையாக உள்ளதாக 2017-ஆம் ஆண்டு நடைபெற்ற கூகுள் கணக்கெடுப்பில் தெரிய வந்தது.
இரண்டாயிரத்து ஐந்நூறு ஆண்டுகளுக்கும் மேல் பழமை வாய்ந்த இலக்கிய மரபைக் கொண்டுள்ள தமிழ் மொழி, தற்போது வழக்கில் இருக்கும் ஒரு சில செம்மொழிகளில் இதுவும் ஒன்றாகும். மேலும் கவனமாகப் பழைய அமைப்புக்களைக் காக்கும் மரபினாலும் பழங்கால இலக்கிய நடை கூட மக்களால் புரிந்து கொள்ளும் நிலையில் உள்ளது. எடுத்துக்காட்டாக, பள்ளிக் குழந்தைகள் சிறுவயதில் கற்கும் அகர வரிசையான ஆத்திசூடி 1,000 ஆண்டுகளுக்கு முன்னர் இயற்றப்பட்டது. திருக்குறள் ஏறத்தாழ 2,000 ஆண்டுகளுக்கு முன் இயற்றப்பட்டது.
##
ஆத்திசூடி எவ்வளவு  ஆண்டுகளுக்கு முன்னர் இயற்றப்பட்டது?
##
1,000
***
தமிழ் தமிழர்களினதும் தமிழ் பேசும் பலரின் தாய்மொழி ஆகும். தமிழ், உலகின் உள்ள முதன்மையான மொழிகளில் ஒன்றும் செம்மொழியும் ஆகும். இந்தியா, இலங்கை, மலேசியா, சிங்கப்பூர் ஆகிய நாடுகளில் அதிக அளவிலும், ஐக்கிய அரபு அமீரகம், தென்னாப்பிரிக்கா, மொரிசியசு, பிசி, இரீயூனியன், திரினிடாடு போன்ற நாடுகளில் சிறிய அளவிலும் தமிழ் பேசப்படுகிறது. 1997-ஆம் ஆண்டுப் புள்ளி விவரப்படி உலகம் முழுவதிலும் 8 கோடி (80 மில்லியன்) மக்களால் பேசப்படும் தமிழ், ஒரு மொழியைத் தாய்மொழியாகக் கொண்டு பேசும் மக்களின் எண்ணிக்கை அடிப்படையில் பதினெட்டாவது இடத்தில் உள்ளது. இணையத்தில் அதிகம் பயன்படுத்தப்படும் இந்திய மொழிகளில் தமிழ் முதன்மையாக உள்ளதாக 2017-ஆம் ஆண்டு நடைபெற்ற கூகுள் கணக்கெடுப்பில் தெரிய வந்தது.
இரண்டாயிரத்து ஐந்நூறு ஆண்டுகளுக்கும் மேல் பழமை வாய்ந்த இலக்கிய மரபைக் கொண்டுள்ள தமிழ் மொழி, தற்போது வழக்கில் இருக்கும் ஒரு சில செம்மொழிகளில் இதுவும் ஒன்றாகும். மேலும் கவனமாகப் பழைய அமைப்புக்களைக் காக்கும் மரபினாலும் பழங்கால இலக்கிய நடை கூட மக்களால் புரிந்து கொள்ளும் நிலையில் உள்ளது. எடுத்துக்காட்டாக, பள்ளிக் குழந்தைகள் சிறுவயதில் கற்கும் அகர வரிசையான ஆத்திசூடி 1,000 ஆண்டுகளுக்கு முன்னர் இயற்றப்பட்டது. திருக்குறள் ஏறத்தாழ 2,000 ஆண்டுகளுக்கு முன் இயற்றப்பட்டது.
##
திருக்குறள் ஏறத்தாழ எவ்வளவு ஆண்டுகளுக்கு முன் இயற்றப்பட்டது?
##
2,000
***
தமிழ்  தமிழ்நாட்டின் ஆட்சி மொழியாகும். அத்துடன் இந்திய அரசியலமைப்பின், எட்டாவது பட்டியலில் குறிப்பிடப்பட்டுள்ள 22 மொழிகளுள் ஒன்றாகவும் உள்ளது. இலங்கையில் மூன்று ஆட்சி மொழிகளுள் தமிழும் ஒன்று. இந்தியாவில் தமிழ்நாடு மாநிலத்திலும் புதுச்சேரி ஒன்றியப் பகுதியிலும் தமிழ் அரச அலுவல் மொழியாக இருக்கிறது. சிங்கப்பூரிலும் நான்கு ஆட்சி மொழிகளுள் தமிழும் ஒன்று. தென்னாப்பிரிக்காவிலும் தமிழுக்கு அரசியலமைப்பு அங்கீகாரம் உள்ளது. மலேசியாவிலும் முதல் நான்கு ஆட்சி மொழிகளில் தமிழும் இடம்பெற்றுள்ளது. மலேசியாவில் தொடக்க இடைநிலைப்பள்ளிகளில் தமிழ் கற்பிக்கப்படுகிறது. நாடு முழுவதும் 523 தமிழ்த் தொடக்கப்பள்ளிகள் அரசுப் பள்ளிகளாக இயங்குகின்றன.
##
தமிழ்  எந்த மாநிலத்தின் ஆட்சி மொழியாகும்?
##
தமிழ்நாட்டின்
***
தமிழ்  தமிழ்நாட்டின் ஆட்சி மொழியாகும். அத்துடன் இந்திய அரசியலமைப்பின், எட்டாவது பட்டியலில் குறிப்பிடப்பட்டுள்ள 22 மொழிகளுள் ஒன்றாகவும் உள்ளது. இலங்கையில் மூன்று ஆட்சி மொழிகளுள் தமிழும் ஒன்று. இந்தியாவில் தமிழ்நாடு மாநிலத்திலும் புதுச்சேரி ஒன்றியப் பகுதியிலும் தமிழ் அரச அலுவல் மொழியாக இருக்கிறது. சிங்கப்பூரிலும் நான்கு ஆட்சி மொழிகளுள் தமிழும் ஒன்று. தென்னாப்பிரிக்காவிலும் தமிழுக்கு அரசியலமைப்பு அங்கீகாரம் உள்ளது. மலேசியாவிலும் முதல் நான்கு ஆட்சி மொழிகளில் தமிழும் இடம்பெற்றுள்ளது. மலேசியாவில் தொடக்க இடைநிலைப்பள்ளிகளில் தமிழ் கற்பிக்கப்படுகிறது. நாடு முழுவதும் 523 தமிழ்த் தொடக்கப்பள்ளிகள் அரசுப் பள்ளிகளாக இயங்குகின்றன.
##
இலங்கையில் எத்தனை  ஆட்சி மொழிகள் உள்ளன ?
##
மூன்று
***
தமிழ்  தமிழ்நாட்டின் ஆட்சி மொழியாகும். அத்துடன் இந்திய அரசியலமைப்பின், எட்டாவது பட்டியலில் குறிப்பிடப்பட்டுள்ள 22 மொழிகளுள் ஒன்றாகவும் உள்ளது. இலங்கையில் மூன்று ஆட்சி மொழிகளுள் தமிழும் ஒன்று. இந்தியாவில் தமிழ்நாடு மாநிலத்திலும் புதுச்சேரி ஒன்றியப் பகுதியிலும் தமிழ் அரச அலுவல் மொழியாக இருக்கிறது. சிங்கப்பூரிலும் நான்கு ஆட்சி மொழிகளுள் தமிழும் ஒன்று. தென்னாப்பிரிக்காவிலும் தமிழுக்கு அரசியலமைப்பு அங்கீகாரம் உள்ளது. மலேசியாவிலும் முதல் நான்கு ஆட்சி மொழிகளில் தமிழும் இடம்பெற்றுள்ளது. மலேசியாவில் தொடக்க இடைநிலைப்பள்ளிகளில் தமிழ் கற்பிக்கப்படுகிறது. நாடு முழுவதும் 523 தமிழ்த் தொடக்கப்பள்ளிகள் அரசுப் பள்ளிகளாக இயங்குகின்றன.
##
சிங்கப்பூரில் எத்தனை  ஆட்சி மொழிகள் உள்ளன ?
##
நான்கு
***
தமிழ்  தமிழ்நாட்டின் ஆட்சி மொழியாகும். அத்துடன் இந்திய அரசியலமைப்பின், எட்டாவது பட்டியலில் குறிப்பிடப்பட்டுள்ள 22 மொழிகளுள் ஒன்றாகவும் உள்ளது. இலங்கையில் மூன்று ஆட்சி மொழிகளுள் தமிழும் ஒன்று. இந்தியாவில் தமிழ்நாடு மாநிலத்திலும் புதுச்சேரி ஒன்றியப் பகுதியிலும் தமிழ் அரச அலுவல் மொழியாக இருக்கிறது. சிங்கப்பூரிலும் நான்கு ஆட்சி மொழிகளுள் தமிழும் ஒன்று. தென்னாப்பிரிக்காவிலும் தமிழுக்கு அரசியலமைப்பு அங்கீகாரம் உள்ளது. மலேசியாவிலும் முதல் நான்கு ஆட்சி மொழிகளில் தமிழும் இடம்பெற்றுள்ளது. மலேசியாவில் தொடக்க இடைநிலைப்பள்ளிகளில் தமிழ் கற்பிக்கப்படுகிறது. நாடு முழுவதும் 523 தமிழ்த் தொடக்கப்பள்ளிகள் அரசுப் பள்ளிகளாக இயங்குகின்றன.
##
நாடு முழுவதும் எத்தனை தமிழ்த் தொடக்கப்பள்ளிகள் அரசுப் பள்ளிகளாக இயங்குகின்றன ?
##
523
***
அலெக்சாந்தர் 16 ஆம் வயதில் தனது கல்வியை அரிஸ்டாட்டிலிடம் பயின்று முடித்தார். பய்சான்டியான் உடனான மன்னர் பிலிப்பின் வாரிசு போட்டியில் வெற்றிபெற்ற அலெக்சாண்டர் வெளிப்படையான ஒரே வாரிசாக அறிவிக்கப்பட்டார். பிலிப் இல்லாத காலகட்டங்களில் திரேசிய நாட்டினர் மாக்கெடோனின் மீது படையெடுத்தனர். அலெக்சாண்டர் உடனே அதற்குத் தக்க பதிலடி கொடுத்தார். அவர்களை தனது எல்லையில் இருந்து ஓட ஓட விரட்டியடித்தார். பின்னர் கிரேக்கத்தில் குடியேற்ற ஆதிக்கத்தை விதைத்தார். அது மட்டுமல்லாது அலெக்சாண்ட்ரோபோலிஸ் என்கிற நகரையும் நிர்மாணித்தார்.பின்னர் பிலிப் திரும்பி வந்ததும் அலெக்சாண்டரை ஒரு சிறு படைக்குத் தலைமையாக நியமித்து அனுப்பி தெற்கு திரேசை கைப்பற்றி வரப் பணித்தார். கிரேக்கத்தின் பெரிந்தஸ் நகரத்திற்கு எதிரான போரில் அலெக்சாண்டர் தனது தந்தையின் உயிரை காப்பாற்றினார். இதை தொடர்ந்து கிரேக்கத்தில் நடந்த பல போர்களில் தனது தந்தையின் வழிகாட்டுதலின் பேரில் நிறைய வெற்றிகளை ஈட்டினார்.பின்னாளில் சில காலம் கழித்து கி.மு.338 ல் அலெக்சாண்டரின் படைகளும் அவரது தந்தை இரண்டாம் பிலிப்பின் படைகளும் இணைந்தன. மேலும் அவை தெற்கு தெர்மொபைலியா வழியாக வலம் வந்தன. பின்னர் தேபான் காரிசன் கூட்டத்திடம் வெற்றியை ஈட்டிய பின்னர் தேபெஸ் மற்றும் ஏதென்சிலிருந்து சிலநாள் பயண தூரத்திலிருந்த ஏலாடிய நகரை வெற்றி பெற்றனர். பின்னர் டிமொஸ்திநீசால் ஆளப்பட்ட ஏதென்சு, தேபெசு மாக்கெடோனியாவை வெல்ல முயன்றனர். அதேநேரம் ஃபிலிப்பும், ஏதென்சும் தேபெஸ் மீது முகாமிட்டு வெல்ல முயன்றனர். ஆனால் இந்தப் போட்டியில் ஏதென்ஸ் வென்றது. பின்னர் ஃபிலிப் அம்பிச்ஸா நோக்கி சென்றார்.
##
அலெக்சாந்தர் 16 ஆம் வயதில் தனது கல்வியை யாரிடம் பயின்று முடித்தார்?
##
அரிஸ்டாட்டிலிடம்
***
அலெக்சாந்தர் 16 ஆம் வயதில் தனது கல்வியை அரிஸ்டாட்டிலிடம் பயின்று முடித்தார். பய்சான்டியான் உடனான மன்னர் பிலிப்பின் வாரிசு போட்டியில் வெற்றிபெற்ற அலெக்சாண்டர் வெளிப்படையான ஒரே வாரிசாக அறிவிக்கப்பட்டார். பிலிப் இல்லாத காலகட்டங்களில் திரேசிய நாட்டினர் மாக்கெடோனின் மீது படையெடுத்தனர். அலெக்சாண்டர் உடனே அதற்குத் தக்க பதிலடி கொடுத்தார். அவர்களை தனது எல்லையில் இருந்து ஓட ஓட விரட்டியடித்தார். பின்னர் கிரேக்கத்தில் குடியேற்ற ஆதிக்கத்தை விதைத்தார். அது மட்டுமல்லாது அலெக்சாண்ட்ரோபோலிஸ் என்கிற நகரையும் நிர்மாணித்தார்.பின்னர் பிலிப் திரும்பி வந்ததும் அலெக்சாண்டரை ஒரு சிறு படைக்குத் தலைமையாக நியமித்து அனுப்பி தெற்கு திரேசை கைப்பற்றி வரப் பணித்தார். கிரேக்கத்தின் பெரிந்தஸ் நகரத்திற்கு எதிரான போரில் அலெக்சாண்டர் தனது தந்தையின் உயிரை காப்பாற்றினார். இதை தொடர்ந்து கிரேக்கத்தில் நடந்த பல போர்களில் தனது தந்தையின் வழிகாட்டுதலின் பேரில் நிறைய வெற்றிகளை ஈட்டினார்.பின்னாளில் சில காலம் கழித்து கி.மு.338 ல் அலெக்சாண்டரின் படைகளும் அவரது தந்தை இரண்டாம் பிலிப்பின் படைகளும் இணைந்தன. மேலும் அவை தெற்கு தெர்மொபைலியா வழியாக வலம் வந்தன. பின்னர் தேபான் காரிசன் கூட்டத்திடம் வெற்றியை ஈட்டிய பின்னர் தேபெஸ் மற்றும் ஏதென்சிலிருந்து சிலநாள் பயண தூரத்திலிருந்த ஏலாடிய நகரை வெற்றி பெற்றனர். பின்னர் டிமொஸ்திநீசால் ஆளப்பட்ட ஏதென்சு, தேபெசு மாக்கெடோனியாவை வெல்ல முயன்றனர். அதேநேரம் ஃபிலிப்பும், ஏதென்சும் தேபெஸ் மீது முகாமிட்டு வெல்ல முயன்றனர். ஆனால் இந்தப் போட்டியில் ஏதென்ஸ் வென்றது. பின்னர் ஃபிலிப் அம்பிச்ஸா நோக்கி சென்றார்.
##
எப்பொழுது  அலெக்சாண்டரின் படைகளும் அவரது தந்தை இரண்டாம் பிலிப்பின் படைகளும் இணைந்தன?
##
கி.மு.338
***
செர்னோபில் அணு உலை விபத்து என்பது 1986 ஆம் ஆண்டு ஏப்ரல் 26 ஆம் நாள் அன்றைய சோவியத் ஒன்றியத்தின் உக்ரேனிலுள்ள பிப்யாட் என்னும் இடத்துக்கு அருகில் அமைந்திருந்த செர்னோபில் அணு உலையில் ஏற்பட்ட விபத்தைக் குறிக்கும்.இங்கு நான்கு அணு உலைகள் இருந்தன. அவற்றில் ஒன்றின் நீர் குளிர்வு சாதனம் செயல்படாததன் காரணமாக, வெப்பம் அதிகரித்து, அணு உலையின் மையம் உருக ஆரம்பித்தது. 1986, ஏப்ரல் 26 அன்று அதிகாலை 1.23 மணியளவில் நேர்ந்த இந்த சம்பவம் தொடர்ந்து மூன்று மணி நேர வெப்பத்தின் காரணமாக , உருகுதலோடு, உலையில் உள்ள பிளாக்குகளைத் தீப்பற்ற செய்து வெடிக்கவும் வைத்தது. அதனால் சுமார் 20 வகையான கதிர் வீச்சுப் பொருள்கள் காற்று மண்டலத்தில் புகுந்தன. காற்று மண்டலத்தில் புகுந்த இக்கதிர் வீச்சுப் பொருள்கள் அங்கிருந்த நிலத்தில் விழுந்து சிதறின. இந்த கதிரியக்க வீழ் பொருள்கள் பத்து ஹிரோஷிமாக் குண்டுகளுக்கு சமமானதாக கருத்தப்படுகின்றது.
செர்னோபில் அணு உலையின் கதிர் வீச்சின் தன்மையை ஆராய பிரித்தானிய அறிவியல் அறிஞர்கள் ஆளில்லா பறக்கும் விமானம் மூலம் சோதனை நடத்துகிறார்கள்.
##
செர்னோபில் அணு உலை விபத்து எப்பொழுது நடந்தது ?
##
1986 ஆம் ஆண்டு ஏப்ரல் 26 ஆம் நாள்
***
செர்னோபில் அணு உலை விபத்து என்பது 1986 ஆம் ஆண்டு ஏப்ரல் 26 ஆம் நாள் அன்றைய சோவியத் ஒன்றியத்தின் உக்ரேனிலுள்ள பிப்யாட் என்னும் இடத்துக்கு அருகில் அமைந்திருந்த செர்னோபில் அணு உலையில் ஏற்பட்ட விபத்தைக் குறிக்கும்.இங்கு நான்கு அணு உலைகள் இருந்தன. அவற்றில் ஒன்றின் நீர் குளிர்வு சாதனம் செயல்படாததன் காரணமாக, வெப்பம் அதிகரித்து, அணு உலையின் மையம் உருக ஆரம்பித்தது. 1986, ஏப்ரல் 26 அன்று அதிகாலை 1.23 மணியளவில் நேர்ந்த இந்த சம்பவம் தொடர்ந்து மூன்று மணி நேர வெப்பத்தின் காரணமாக , உருகுதலோடு, உலையில் உள்ள பிளாக்குகளைத் தீப்பற்ற செய்து வெடிக்கவும் வைத்தது. அதனால் சுமார் 20 வகையான கதிர் வீச்சுப் பொருள்கள் காற்று மண்டலத்தில் புகுந்தன. காற்று மண்டலத்தில் புகுந்த இக்கதிர் வீச்சுப் பொருள்கள் அங்கிருந்த நிலத்தில் விழுந்து சிதறின. இந்த கதிரியக்க வீழ் பொருள்கள் பத்து ஹிரோஷிமாக் குண்டுகளுக்கு சமமானதாக கருத்தப்படுகின்றது.
செர்னோபில் அணு உலையின் கதிர் வீச்சின் தன்மையை ஆராய பிரித்தானிய அறிவியல் அறிஞர்கள் ஆளில்லா பறக்கும் விமானம் மூலம் சோதனை நடத்துகிறார்கள்.
##
செர்னோபில் அணு உலையில்  எத்தனை அணு உலைகள் இருந்தன?
##
நான்கு
***
போபால் பேரழிவு  டிசம்பர் 3, 1984 ல் இந்தியாவில் உள்ள போபாலில் உள்ள தொழிற்சாலையில் ஏற்பட்ட நச்சு வளிமக் கசிவினால் (வாயுக் கசிவினால்) ஏற்பட்ட பல்லாயிரக்கணக்கான உயிரிழப்புகளை நினைவுகூறும் ஒரு துன்ப நிகழ்வாகும்.யூனியன் கார்பைடு எனும் பூச்சிகொல்லி மருந்து தயாரிக்கும் தொழிற்சாலையில் மீத்தைல் ஐசோ சயனேட் எனும் நச்சு வளிமம் கசிந்ததினால் ஐந்து இலட்சத்திற்கும் மேற்பட்ட மக்கள் பாதிக்கப்பட்டனர். உடனடி உயிரிழப்பாக ஏறத்தாழ 2,259 பேர் நச்சு வளிமம் தாக்கி இறந்தனர். அதற்கடுத்த இரண்டு வாரங்களில் மேலும் 8,000 பேர் இறந்தனர். இன்னும் 8,000 பேர் வளிமத்தின் தாக்கத்தினால் ஏற்பட்ட நோயால் பாதிக்கப்பட்டு இறந்தனர். போப்பால் பேரழிவு உலகில் உள்ள தொழிற்சாலைகளால் ஏற்பட்ட பேரழிவுகளில் மிக அதிகப் பாதிப்பை ஏற்படுத்திய பேரழிவாகக் கருதப்படுகிறது. இதனால் அங்குள்ள பாதிப்புகளை ஆராய 1993 ஆம் ஆண்டு அனைத்து நாடு மருத்துவக்குழு ஆணையம் இங்கு ஏற்படுத்தப்பட்டது.இந்த நிகழ்வுக்குக் காரணமான யூனியன் கார்பைடு இந்தியா லிமிடட் என்ற நிறுவனம் 51% உரிமையுடன் 1969 ல் போப்பாலில் நிறுவப்பட்டதாகும். இதன் உரிமை யூனியன் கார்பைடு கார்ப்பொரேசனுக்கு சொந்தமானதாகும். இதன் 49% உரிமை இந்திய நிர்வாகத்திற்குச் சொந்தமானதாகும். இந்நிறுவனத்தின் முக்கிய குற்றவாளியாக சித்தரிக்கப்பட்ட யூனியன் கார்பைடு நிறுவனத்தின் அப்போதைய முதன்மை செயல் அதிகாரி வாரன் அண்டர்சன் இந்த வழக்கில் குற்றவாளியாக அறிவிக்கப்பட்டார். விபத்து நடந்தபின் இந்தியாவில் வைத்துக் கைது செய்யப்பட்ட வாரன் அண்டர்சன், அப்போதைய அரசியல் தலையீடுகளால் இந்தியாவை விட்டு கௌரவத்தோடு விமானத்தில் ஏற்றி அவரது தாய்நாட்டிற்கு அனுப்பி வைக்கப்பட்டார்.பிணையம் பெற்று வெளிவந்த அண்டர்சன் அமெரிக்காவிற்கு திரும்பிய பின் மீண்டும் இந்தியா வர மறுத்தார். அமெரிக்கக் குடிமகனான ஆண்டர்சனை இந்தியாவிடம் ஒப்படைக்க அமெரிக்க அரசு மறுத்து வந்தது.செப்டம்பர் 29, 2014 அன்று அமெரிக்காவில் இவர் இறந்தார்.
##
போபால் பேரழிவு எப்பொழுது நடந்தது?
##
டிசம்பர் 3, 1984
***
போபால் பேரழிவு  டிசம்பர் 3, 1984 ல் இந்தியாவில் உள்ள போபாலில் உள்ள தொழிற்சாலையில் ஏற்பட்ட நச்சு வளிமக் கசிவினால் (வாயுக் கசிவினால்) ஏற்பட்ட பல்லாயிரக்கணக்கான உயிரிழப்புகளை நினைவுகூறும் ஒரு துன்ப நிகழ்வாகும்.யூனியன் கார்பைடு எனும் பூச்சிகொல்லி மருந்து தயாரிக்கும் தொழிற்சாலையில் மீத்தைல் ஐசோ சயனேட் எனும் நச்சு வளிமம் கசிந்ததினால் ஐந்து இலட்சத்திற்கும் மேற்பட்ட மக்கள் பாதிக்கப்பட்டனர். உடனடி உயிரிழப்பாக ஏறத்தாழ 2,259 பேர் நச்சு வளிமம் தாக்கி இறந்தனர். அதற்கடுத்த இரண்டு வாரங்களில் மேலும் 8,000 பேர் இறந்தனர். இன்னும் 8,000 பேர் வளிமத்தின் தாக்கத்தினால் ஏற்பட்ட நோயால் பாதிக்கப்பட்டு இறந்தனர். போப்பால் பேரழிவு உலகில் உள்ள தொழிற்சாலைகளால் ஏற்பட்ட பேரழிவுகளில் மிக அதிகப் பாதிப்பை ஏற்படுத்திய பேரழிவாகக் கருதப்படுகிறது. இதனால் அங்குள்ள பாதிப்புகளை ஆராய 1993 ஆம் ஆண்டு அனைத்து நாடு மருத்துவக்குழு ஆணையம் இங்கு ஏற்படுத்தப்பட்டது.இந்த நிகழ்வுக்குக் காரணமான யூனியன் கார்பைடு இந்தியா லிமிடட் என்ற நிறுவனம் 51% உரிமையுடன் 1969 ல் போப்பாலில் நிறுவப்பட்டதாகும். இதன் உரிமை யூனியன் கார்பைடு கார்ப்பொரேசனுக்கு சொந்தமானதாகும். இதன் 49% உரிமை இந்திய நிர்வாகத்திற்குச் சொந்தமானதாகும். இந்நிறுவனத்தின் முக்கிய குற்றவாளியாக சித்தரிக்கப்பட்ட யூனியன் கார்பைடு நிறுவனத்தின் அப்போதைய முதன்மை செயல் அதிகாரி வாரன் அண்டர்சன் இந்த வழக்கில் குற்றவாளியாக அறிவிக்கப்பட்டார். விபத்து நடந்தபின் இந்தியாவில் வைத்துக் கைது செய்யப்பட்ட வாரன் அண்டர்சன், அப்போதைய அரசியல் தலையீடுகளால் இந்தியாவை விட்டு கௌரவத்தோடு விமானத்தில் ஏற்றி அவரது தாய்நாட்டிற்கு அனுப்பி வைக்கப்பட்டார்.பிணையம் பெற்று வெளிவந்த அண்டர்சன் அமெரிக்காவிற்கு திரும்பிய பின் மீண்டும் இந்தியா வர மறுத்தார். அமெரிக்கக் குடிமகனான ஆண்டர்சனை இந்தியாவிடம் ஒப்படைக்க அமெரிக்க அரசு மறுத்து வந்தது.செப்டம்பர் 29, 2014 அன்று அமெரிக்காவில் இவர் இறந்தார்.
##
வாரன் அண்டர்சன் எப்பொழுது  இறந்தார்?
##
செப்டம்பர் 29, 2014
***
மார்வெல் திரைப் பிரபஞ்சத்தின் முதலாம் கட்டமாக 2008 இல் அயன் மேன் வெளியீட்டில் தொடங்கி 2012 இல் தி அவேஞ்சர்ஸ் வெளியீட்டில் முடிந்தது. மூன்றாம் கட்டமாக 2016 இல் கேப்டன் அமெரிக்கா: சிவில் வார் வெளியீட்டில் தொடங்கி 2019 இல் இசுபைடர்-மேன்: பார் பிரம் ஹோம் வெளியீட்டில் முடிந்தது. இந்த மூன்று கட்டங்களையும் கூட்டாக தி இன்பினிட்டி சாகா என்று அழைக்கப்படுகின்றது. நான்காம் கட்டமாக 2021 இல் பிளாக் விடோ உடன் தொடங்குகிறது.
##
அயன் மேன் திரைப்படம் எப்பொழுது வெளியானது ?
##
2008
***
முதல் டிசி நீட்டிக்கப்பட்ட பிரபஞ்சத்தின் படமாக மேன் ஆப் ஸ்டீல் என்ற திரைப்படம் 2013 ஆம் ஆண்டு வெளியானது. இது சூப்பர் மேன் திரைப்படத்தின் மீள் உருவாக்கம் ஆகும். அதை தொடர்ந்து பேட்மேன் வி சூப்பர்மேன்: டவுன் ஆஃப் ஜஸ்டிஸ் (2016), சூசைட் ஸ்க்வாட் (2016), வொண்டர் வுமன் (2017), ஜஸ்டிஸ் லீக் (2017), அக்குவாமேன்[6] (2017), ஷசாம்! (2019), பேர்ட்ஸ் ஆஃப் பிரே (2020), வொண்டர் வுமன் 1984 (2020) போன்ற திரைப்படங்கள் வெளியானது.
##
டிசி நீட்டிக்கப்பட்ட பிரபஞ்சத்தின் முதல் படம் எது?
##
மேன் ஆப் ஸ்டீல்
***
சந்திரயான்-1  என்பது இந்திய விண்வெளி ஆய்வு மையத்தால் 2008, அக்டோபர் 22 இல் விண்வெளிக்குச் செலுத்தப்பட்ட ஆளில்லாத நிலவுப் பயணம் ஆகும்[1]. இதன் முக்கிய நோக்கம் நிலவுப்பரப்பில் பல்வேறு தாதுக்கள் மற்றும் வேதிமூலகங்களின் பரவலை ஆய்வு செய்வதும், முழு நிலவுப் பரப்பையும் அதிக துல்லியத்துடன் முப்பரிமாண வரைபடமாக்கலும் ஆகும். இந்திய விண்வெளி ஆய்வு நிறுவனத்தின் துருவ செயற்கைக்கோள் ஏவுவாகனமான பி.எஸ்.எல்.வி. சந்திராயன் I கலத்தை 240 கி.மீ x 24000 கி.மீ புவிச் சுற்றுப்பாதையில் செலுத்தும். பின்னர் விண்கலமானது தன்னகத்துள்ள முன்னுந்து அமைப்பின் துணைகொண்டு நிலவைச்சுற்றிய 100 கி.மீ துருவச் சுற்றுப்பாதையில் நிலைநிறுத்தப்படும். சந்திராயன் I விண்கலமானது சுற்றிவரக்கூடிய அமைப்பையும் நிலவில் இறங்கக்கூடிய அமைப்பையும் ஒருங்கே கொண்டிருக்கும்.இப்பணித்திட்டத்தின் தலைவராக மயில்சாமி அண்ணாதுரை இருந்து வருகிறார். இத்திட்டத்திற்கு இந்திய ரூபாயில் 3.8 பில்லியன் (சுமார் 83 மில்லியன் அமெரிக்க டாலர்கள்) செலவாகும் என்று மதிப்பிடப்பட்டுள்ளது.
##
சந்திரயான்-1 எப்பொழுது  விண்வெளிக்குச் செலுத்தப்பத்தது ?
##
2008, அக்டோபர் 22
***
சந்திரயான் இரண்டு ஆண்டுகள் செயல்படும் என்று அறிவிக்கப்பட்டிருந்தது. ஆனால் விண்ணில் செலுத்தப்பட்ட 312 நாட்களுக்கு பிறகு எதிர்பாராத விதமாக சந்திரயானின் தரைக்கட்டுப்பாடு நிலையத்துடனான இணைப்பு துண்டித்தது. மைக்ரோ ஒளிக்கற்றை மூலம் நாசா விஞ்ஞானிகள் நடத்திய ஆய்வில் சந்திரயான் செயலிழந்த நிலையில் நிலவுக்கு மேலே 200 கி.மீ தொலைவில் சுற்றிக்கொண்டிருப்பது கண்டுபிடிக்கப்பட்டது.10 மாதங்களே செயல்பட்டாலும் சந்திரயான் தனது திட்ட நோக்கத்தில் 95 சதவீதத்தை வெற்றிகரமாக நிறைவு செய்ததாக விஞ்ஞானிகள் அறிவித்தனர்.
##
சந்திரயான் எத்தனை  ஆண்டுகள் செயல்படும் என்று அறிவிக்கப்பட்டிருந்தது.
##
இரண்டு
***
இந்தியாவின் முதல் செயற்கைக்கோள் ஆரியபட்டா  ஏப்ரல் 19 , 1975 அன்று விண்ணில் ஏவப்பட்டது. இதனைத் தொடர்ந்து ரோகினி வகை செயற்கைக்கோள்களை இந்தியாவிலேயே தயாரித்து ஏவுதலும் நிகழ்ந்தது. தற்போது இஃச்ரோ பல்வகையான புவி கூர்நோக்கு செயற்கைக்கோள்களை இயக்கி வருகிறது.
##
இந்தியாவின் முதல் செயற்கைக்கோள் எது ?
##
ஆரியபட்டா
***
12 கோபக்கார மனிதர்கள்  1957இல் வெளியிடப்பட்ட ஆங்கிலத் திரைப்படம். இது மிகவும் பாரட்டப்பட்ட, இன்றும் போற்றப்படும் ஒரு திரைப்படமாகும். தொலைக்காட்சித் தொடராகத் தயாரிக்கப்பட்டு பின் திரைக்கு வந்த இந்தப்படம் முழுவதும் ஒரு சிறு அறையில் நடைபெறுவதாக இருக்கும்.1957இல் படம் வெளியிடப்பட்டவுடன் மக்களிடையேயும் பத்திரிக்கைகளிலும் இந்தப்படம் மிகப்பெறும் வரவேற்பைப் பெற்றது. 2007இல், இந்தப் படம், அமெரிக்காவின் தேசிய படக் காப்பகத்தால் லைப்ரரி ஆஃப் காங்கிரஸில் பாதுகாக்கப்படும் பட்டியலில் சேர்க்கப்பட்டது.  இப்படம் ராட்டன் டொமேட்டோஸ்  தளத்தின் விமர்சனம் மற்றும் வாசகர் ஏற்பில் 100% மதிப்பு பெற்றுள்ளது.
##
12 கோபக்கார மனிதர்கள் எம்மொழி திரைப்படம்?
##
ஆங்கிலத் திரைப்படம்
***
12 கோபக்கார மனிதர்கள்  1957இல் வெளியிடப்பட்ட ஆங்கிலத் திரைப்படம். இது மிகவும் பாரட்டப்பட்ட, இன்றும் போற்றப்படும் ஒரு திரைப்படமாகும். தொலைக்காட்சித் தொடராகத் தயாரிக்கப்பட்டு பின் திரைக்கு வந்த இந்தப்படம் முழுவதும் ஒரு சிறு அறையில் நடைபெறுவதாக இருக்கும்.1957இல் படம் வெளியிடப்பட்டவுடன் மக்களிடையேயும் பத்திரிக்கைகளிலும் இந்தப்படம் மிகப்பெறும் வரவேற்பைப் பெற்றது. 2007இல், இந்தப் படம், அமெரிக்காவின் தேசிய படக் காப்பகத்தால் லைப்ரரி ஆஃப் காங்கிரஸில் பாதுகாக்கப்படும் பட்டியலில் சேர்க்கப்பட்டது.  இப்படம் ராட்டன் டொமேட்டோஸ்  தளத்தின் விமர்சனம் மற்றும் வாசகர் ஏற்பில் 100% மதிப்பு பெற்றுள்ளது.
##
12 கோபக்கார மனிதர்கள் திரைப்படம் ராட்டன் டொமேட்டோஸ்  தளத்தின் விமர்சனம் மற்றும் வாசகர் ஏற்பில் எத்தனை % மதிப்பு பெற்றுள்ளது?
##
100%

***
இந்தியாவின் தேசிய வருமானத்தில் முக்கிய பங்கு வகிப்பது  வேளாண்மை.மிக அதிக நீளமான கடற்கரையைக் கொண்ட தென் மாநிலம் ஆந்திரப்பிரதேசம்.
ஈராக் நாட்டின் தலைநகரம் பாக்தாக்.இந்திய அறிவயற் கழகம் அமைதுள்ள நகரம் பெங்களூர்.இந்தியா முதல் அணுகுண்டு சோதனை நடத்திய இடம் பொகரான்.
##
இந்தியாவின் தேசிய வருமானத்தில் முக்கிய பங்கு வகிப்பது
##
வேளாண்மை
***
இந்தியாவின் தேசிய வருமானத்தில் முக்கிய பங்கு வகிப்பது  வேளாண்மை.மிக அதிக நீளமான கடற்கரையைக் கொண்ட தென் மாநிலம் ஆந்திரப்பிரதேசம்.
ஈராக் நாட்டின் தலைநகரம் பாக்தாக்.இந்திய அறிவயற் கழகம் அமைதுள்ள நகரம் பெங்களூர்.இந்தியா முதல் அணுகுண்டு சோதனை நடத்திய இடம் பொகரான்.
##
மிக அதிக நீளமான கடற்கரையைக் கொண்ட தென் மாநிலம் எது?
##
ஆந்திரப்பிரதேசம்
***
இந்தியாவின் தேசிய வருமானத்தில் முக்கிய பங்கு வகிப்பது  வேளாண்மை.மிக அதிக நீளமான கடற்கரையைக் கொண்ட தென் மாநிலம் ஆந்திரப்பிரதேசம்.
ஈராக் நாட்டின் தலைநகரம் பாக்தாக்.இந்திய அறிவயற் கழகம் அமைதுள்ள நகரம் பெங்களூர்.இந்தியா முதல் அணுகுண்டு சோதனை நடத்திய இடம் பொகரான்.
##
ஈராக் நாட்டின் தலைநகரம்?
##
பாக்தாக்
***
இந்தியாவின் தேசிய வருமானத்தில் முக்கிய பங்கு வகிப்பது  வேளாண்மை.மிக அதிக நீளமான கடற்கரையைக் கொண்ட தென் மாநிலம் ஆந்திரப்பிரதேசம்.
ஈராக் நாட்டின் தலைநகரம் பாக்தாக்.இந்திய அறிவயற் கழகம் அமைதுள்ள நகரம் பெங்களூர்.இந்தியா முதல் அணுகுண்டு சோதனை நடத்திய இடம் பொகரான்.
##
இந்திய அறிவயற் கழகம் அமைதுள்ள நகரம்?
##
பெங்களூர்
***
இந்தியாவின் தேசிய வருமானத்தில் முக்கிய பங்கு வகிப்பது  வேளாண்மை.மிக அதிக நீளமான கடற்கரையைக் கொண்ட தென் மாநிலம் ஆந்திரப்பிரதேசம்.
ஈராக் நாட்டின் தலைநகரம் பாக்தாக்.இந்திய அறிவயற் கழகம் அமைதுள்ள நகரம் பெங்களூர்.இந்தியா முதல் அணுகுண்டு சோதனை நடத்திய இடம் பொகரான்.
##
இந்தியா முதல் அணுகுண்டு சோதனை நடத்திய இடம்?
##
பொகரான்
***
சின்னசாமி விளையாட்டு அரங்கம் பெங்களூர் ஊரில் அமைந்துள்ளது.இந்தியாவின் தேசிய விளையாட்டு ஹாக்கி.
ஹாக்கி விளையாட்டில் ஒர் அணியில் ஆடுவோர் எண்ணிக்கை 11.போலோ விளையாட்டிற்கு மிகப்பெரிய மைதானம் தேவை.
விம்பிள்டன் டென்னிஸ் விளையாட்டுடன் தொடர்புடையது.
##
சின்னசாமி விளையாட்டு அரங்கம் எந்த ஊரில் அமைந்துள்ளது?
##
பெங்களூர்
***
சின்னசாமி விளையாட்டு அரங்கம் பெங்களூர் ஊரில் அமைந்துள்ளது.இந்தியாவின் தேசிய விளையாட்டு ஹாக்கி.
ஹாக்கி விளையாட்டில் ஒர் அணியில் ஆடுவோர் எண்ணிக்கை 11.போலோ விளையாட்டிற்கு மிகப்பெரிய மைதானம் தேவை.
விம்பிள்டன் டென்னிஸ் விளையாட்டுடன் தொடர்புடையது.
##
இந்தியாவின் தேசிய விளையாட்டு எது?
##
ஹாக்கி
***
சின்னசாமி விளையாட்டு அரங்கம் பெங்களூர் ஊரில் அமைந்துள்ளது.இந்தியாவின் தேசிய விளையாட்டு ஹாக்கி.
ஹாக்கி விளையாட்டில் ஒர் அணியில் ஆடுவோர் எண்ணிக்கை 11.போலோ விளையாட்டிற்கு மிகப்பெரிய மைதானம் தேவை.
விம்பிள்டன் டென்னிஸ் விளையாட்டுடன் தொடர்புடையது.
##
ஹாக்கி விளையாட்டில் ஒர் அணியில் ஆடுவோர் எண்ணிக்கை எத்தனை?
##
11
***
சின்னசாமி விளையாட்டு அரங்கம் பெங்களூர் ஊரில் அமைந்துள்ளது.இந்தியாவின் தேசிய விளையாட்டு ஹாக்கி.
ஹாக்கி விளையாட்டில் ஒர் அணியில் ஆடுவோர் எண்ணிக்கை 11.போலோ விளையாட்டிற்கு மிகப்பெரிய மைதானம் தேவை.
விம்பிள்டன் டென்னிஸ் விளையாட்டுடன் தொடர்புடையது.
##
எந்த விளையாட்டிற்கு மிகப்பெரிய மைதானம் தேவை?
##
போலோ
***
சின்னசாமி விளையாட்டு அரங்கம் பெங்களூர் ஊரில் அமைந்துள்ளது.இந்தியாவின் தேசிய விளையாட்டு ஹாக்கி.
ஹாக்கி விளையாட்டில் ஒர் அணியில் ஆடுவோர் எண்ணிக்கை 11.போலோ விளையாட்டிற்கு மிகப்பெரிய மைதானம் தேவை.
விம்பிள்டன் டென்னிஸ் விளையாட்டுடன் தொடர்புடையது.
##
விம்பிள்டன் எந்த விளையாட்டுடன் தொடர்புடையது?
##
டென்னிஸ்
***
ரங்கசாமி கப் ஹாக்கி விளையாட்டுடன் தொடர்புடையது.
பதினோறாவது ஆசிய விளையாட்டுப் போட்டிகள் பெய்ஜிங் நடைபெற்றன.நான்கு வருடம் ஆண்டுகளுக்கு ஒரு முறை ஆசியா விளையாட்டுப் போட்டிகள் நடைபெறுகிறது.கூடைப்பந்து விளையாட்டில் ஒர் அணியில் ஆடுவோர் எண்ணிக்கை 5.
ரஞ்சிக் கோப்பையுடன் தொடர் புடையது கிரிக்கெட் விளையாட்டு.
##
ரங்கசாமி கப் எந்த விளையாட்டுடன் தொடர்புடையது?
##
ஹாக்கி
***
ரங்கசாமி கப் ஹாக்கி விளையாட்டுடன் தொடர்புடையது.
பதினோறாவது ஆசிய விளையாட்டுப் போட்டிகள் பெய்ஜிங் நடைபெற்றன.நான்கு வருடம் ஆண்டுகளுக்கு ஒரு முறை ஆசியா விளையாட்டுப் போட்டிகள் நடைபெறுகிறது.கூடைப்பந்து விளையாட்டில் ஒர் அணியில் ஆடுவோர் எண்ணிக்கை 5.
##
பதினோறாவது ஆசிய விளையாட்டுப் போட்டிகள் எங்கு நடைபெற்றன?
##
பெய்ஜிங்
***
ரங்கசாமி கப் ஹாக்கி விளையாட்டுடன் தொடர்புடையது.
பதினோறாவது ஆசிய விளையாட்டுப் போட்டிகள் பெய்ஜிங் நடைபெற்றன.நான்கு வருடம் ஆண்டுகளுக்கு ஒரு முறை ஆசியா விளையாட்டுப் போட்டிகள் நடைபெறுகிறது.கூடைப்பந்து விளையாட்டில் ஒர் அணியில் ஆடுவோர் எண்ணிக்கை 5.
##
எத்தனன ஆண்டுகளுக்கு ஒரு முறை ஆசியா விளையாட்டுப் போட்டிகள் நடைபெறுகிறது?
##
நான்கு வருடம்
***
ரங்கசாமி கப் ஹாக்கி விளையாட்டுடன் தொடர்புடையது.
பதினோறாவது ஆசிய விளையாட்டுப் போட்டிகள் பெய்ஜிங் நடைபெற்றன.நான்கு வருடம் ஆண்டுகளுக்கு ஒரு முறை ஆசியா விளையாட்டுப் போட்டிகள் நடைபெறுகிறது.கூடைப்பந்து விளையாட்டில் ஒர் அணியில் ஆடுவோர் எண்ணிக்கை 5.
##
கூடைப்பந்து விளையாட்டில் ஒர் அணியில் ஆடுவோர் எண்ணிக்கை எத்தனை?
##
5
***
ரங்கசாமி கப் ஹாக்கி விளையாட்டுடன் தொடர்புடையது.
பதினோறாவது ஆசிய விளையாட்டுப் போட்டிகள் பெய்ஜிங் நடைபெற்றன.நான்கு வருடம் ஆண்டுகளுக்கு ஒரு முறை ஆசியா விளையாட்டுப் போட்டிகள் நடைபெறுகிறது.கூடைப்பந்து விளையாட்டில் ஒர் அணியில் ஆடுவோர் எண்ணிக்கை 5.
ரஞ்சிக் கோப்பையுடன் தொடர் புடையது கிரிக்கெட் .
##
ரஞ்சிக் கோப்பையுடன் தொடர் புடையது எந்த விளையாட்டு?
##
கிரிக்கெட்
***
1982ம் ஆண்டு நடைபெற்ற ஆசிய விளையாட்டின் அடையாளச் சின்னம் அப்பு.சையது மோதி பாட்மின்டன் விளையாட்டுடன் தொடர் புடையவர்.ரன்ஸ் அண்ட் ரூபின்ஸ் நூலை எழுதியவர் கவாஸ்கர்.
கபாடி விளையாட்டின் ஒர் அணியில் ஆடுவோர் எண்ணிக்கை 7.முதல் தெற்கு ஆசிய பெடரேசன் விளையாட்டுப் போட்டி நடைப்பெற்ற இடம் கொழும்பு
##
1982ம் ஆண்டு நடைபெற்ற ஆசிய விளையாட்டின் அடையாளச் சின்னம் எது?
##
அப்பு
***
1982ம் ஆண்டு நடைபெற்ற ஆசிய விளையாட்டின் அடையாளச் சின்னம் அப்பு.சையது மோதி பாட்மின்டன் விளையாட்டுடன் தொடர் புடையவர்.ரன்ஸ் அண்ட் ரூபின்ஸ் நூலை எழுதியவர் கவாஸ்கர்.
கபாடி விளையாட்டின் ஒர் அணியில் ஆடுவோர் எண்ணிக்கை 7.முதல் தெற்கு ஆசிய பெடரேசன் விளையாட்டுப் போட்டி நடைப்பெற்ற இடம் கொழும்பு
##
சையது மோதி எந்த விளையாட்டுடன் தொடர் புடையவர்?
##
பாட்மின்டன்
***
1982ம் ஆண்டு நடைபெற்ற ஆசிய விளையாட்டின் அடையாளச் சின்னம் அப்பு.சையது மோதி பாட்மின்டன் விளையாட்டுடன் தொடர் புடையவர்.ரன்ஸ் அண்ட் ரூபின்ஸ் நூலை எழுதியவர் கவாஸ்கர்.கபாடி விளையாட்டின் ஒர் அணியில் ஆடுவோர் எண்ணிக்கை 7.முதல் தெற்கு ஆசிய பெடரேசன் விளையாட்டுப் போட்டி நடைப்பெற்ற இடம் கொழும்பு.
##
ரன்ஸ் அண்ட் ரூபின்ஸ் நூலை எழுதியவர்?
##
கவாஸ்கர்
***
1982ம் ஆண்டு நடைபெற்ற ஆசிய விளையாட்டின் அடையாளச் சின்னம் அப்பு.சையது மோதி பாட்மின்டன் விளையாட்டுடன் தொடர் புடையவர்.ரன்ஸ் அண்ட் ரூபின்ஸ் நூலை எழுதியவர் கவாஸ்கர்.கபாடி விளையாட்டின் ஒர் அணியில் ஆடுவோர் எண்ணிக்கை 7.முதல் தெற்கு ஆசிய பெடரேசன் விளையாட்டுப் போட்டி நடைப்பெற்ற இடம் கொழும்பு.
##
கபாடி விளையாட்டின் ஒர் அணியில் ஆடுவோர் எண்ணிக்கை எத்தனை?
##
7
***
1982ம் ஆண்டு நடைபெற்ற ஆசிய விளையாட்டின் அடையாளச் சின்னம் அப்பு.சையது மோதி பாட்மின்டன் விளையாட்டுடன் தொடர் புடையவர்.ரன்ஸ் அண்ட் ரூபின்ஸ் நூலை எழுதியவர் கவாஸ்கர்.கபாடி விளையாட்டின் ஒர் அணியில் ஆடுவோர் எண்ணிக்கை 7.முதல் தெற்கு ஆசிய பெடரேசன் விளையாட்டுப் போட்டி நடைப்பெற்ற இடம் கொழும்பு.
##
முதல் தெற்கு ஆசிய பெடரேசன் விளையாட்டுப் போட்டி நடைப்பெற்ற இடம்?
##
கொழும்பு
***
இந்தியா விண்வெளி யுகத்திற்குள் நுழைந்ததற்குக் காரணமானவர் அப்துல் கலாம்.
ஜாலியன் வாலாபாக் படுகொலை நடந்த ஆண்டு 1919.ஆஸ்கார் பரிசு பெற்ற முதல் இந்தியர் சத்யஜித்ரே.
பூமி ஏறத்தாழ கோள வடிவமானது தாலமி முதன்முதலில் கூறியவர்.குஜராத் மாநிலத்தின் தலைநகரம் காந்தி நகர்.
##
இந்தியா விண்வெளி யுகத்திற்குள் நுழைந்ததற்குக் காரணமானவர் ?
##
அப்துல் கலாம்
***
இந்தியா விண்வெளி யுகத்திற்குள் நுழைந்ததற்குக் காரணமானவர் அப்துல் கலாம்.
ஜாலியன் வாலாபாக் படுகொலை நடந்த ஆண்டு 1919.ஆஸ்கார் பரிசு பெற்ற முதல் இந்தியர் சத்யஜித்ரே.பூமி ஏறத்தாழ கோள வடிவமானது தாலமி முதன்முதலில் கூறியவர்.குஜராத் மாநிலத்தின் தலைநகரம் காந்தி நகர்.
##
ஜாலியன் வாலாபாக் படுகொலை நடந்த ஆண்டு
##
1919
***
இந்தியா விண்வெளி யுகத்திற்குள் நுழைந்ததற்குக் காரணமானவர் அப்துல் கலாம்.
ஜாலியன் வாலாபாக் படுகொலை நடந்த ஆண்டு 1919.ஆஸ்கார் பரிசு பெற்ற முதல் இந்தியர் சத்யஜித்ரே.பூமி ஏறத்தாழ கோள வடிவமானது தாலமி முதன்முதலில் கூறியவர்.குஜராத் மாநிலத்தின் தலைநகரம் காந்தி நகர்.
##
ஆஸ்கார் பரிசு பெற்ற முதல் இந்தியர்
##
சத்யஜித்ரே
***
இந்தியா விண்வெளி யுகத்திற்குள் நுழைந்ததற்குக் காரணமானவர் அப்துல் கலாம்.
ஜாலியன் வாலாபாக் படுகொலை நடந்த ஆண்டு 1919.ஆஸ்கார் பரிசு பெற்ற முதல் இந்தியர் சத்யஜித்ரே.பூமி ஏறத்தாழ கோள வடிவமானது தாலமி முதன்முதலில் கூறியவர்.குஜராத் மாநிலத்தின் தலைநகரம் காந்தி நகர்.
##
பூமி ஏறத்தாழ கோள வடிவமானது என்று முதன்முதலில் கூறியவர்?
##
தாலமி
***
இந்தியா விண்வெளி யுகத்திற்குள் நுழைந்ததற்குக் காரணமானவர் அப்துல் கலாம்.
ஜாலியன் வாலாபாக் படுகொலை நடந்த ஆண்டு 1919.ஆஸ்கார் பரிசு பெற்ற முதல் இந்தியர் சத்யஜித்ரே.பூமி ஏறத்தாழ கோள வடிவமானது தாலமி முதன்முதலில் கூறியவர்.குஜராத் மாநிலத்தின் தலைநகரம் காந்தி நகர்.
##
குஜராத் மாநிலத்தின் தலைநகரம்?
##
காந்தி நகர்
***
இந்தியாவின் மிக உயர்ந்த இலக்கிய விருது ஞானபீட விருது.
இந்தியாவில் ராஜஸ்தான் அதிகமாக செம்பு கிடைக்கிறது.
அதிக மக்கள் அடர்த்தியை கொண்ட கண்டம் ஐரோப்பா.
தேசிய காளான் ஆராய்ச்சி மற்றும் பயிற்சி நிலையம் அமைந்துள்ள இடம் சோலன்.உலக அளவில் அதிகமாக பேசப்படும் முதல் மொழி மாண்டரின்.
##
இந்தியாவின் மிக உயர்ந்த இலக்கிய விருது எது?
##
ஞானபீட விருது
***
இந்தியாவின் மிக உயர்ந்த இலக்கிய விருது ஞானபீட விருது.
இந்தியாவில் ராஜஸ்தான் அதிகமாக செம்பு கிடைக்கிறது.
அதிக மக்கள் அடர்த்தியை கொண்ட கண்டம் ஐரோப்பா.
தேசிய காளான் ஆராய்ச்சி மற்றும் பயிற்சி நிலையம் அமைந்துள்ள இடம் சோலன்.உலக அளவில் அதிகமாக பேசப்படும் முதல் மொழி மாண்டரின்.
##
இந்தியாவில் எங்கு அதிகமாக செம்பு கிடைக்கிறது?
##
ராஜஸ்தான்
***
இந்தியாவின் மிக உயர்ந்த இலக்கிய விருது ஞானபீட விருது.
இந்தியாவில் ராஜஸ்தான் அதிகமாக செம்பு கிடைக்கிறது.
அதிக மக்கள் அடர்த்தியை கொண்ட கண்டம் ஐரோப்பா.
தேசிய காளான் ஆராய்ச்சி மற்றும் பயிற்சி நிலையம் அமைந்துள்ள இடம் சோலன்.உலக அளவில் அதிகமாக பேசப்படும் முதல் மொழி மாண்டரின்.
##
அதிக மக்கள் அடர்த்தியை கொண்ட கண்டம் எது?
##
ஐரோப்பா
***
இந்தியாவின் மிக உயர்ந்த இலக்கிய விருது ஞானபீட விருது.
இந்தியாவில் ராஜஸ்தான் அதிகமாக செம்பு கிடைக்கிறது.
அதிக மக்கள் அடர்த்தியை கொண்ட கண்டம் ஐரோப்பா.
தேசிய காளான் ஆராய்ச்சி மற்றும் பயிற்சி நிலையம் அமைந்துள்ள இடம் சோலன்.உலக அளவில் அதிகமாக பேசப்படும் முதல் மொழி மாண்டரின்.
##
தேசிய காளான் ஆராய்ச்சி மற்றும் பயிற்சி நிலையம் அமைந்துள்ள இடம் எது?
##
சோலன்
***
இந்தியாவின் மிக உயர்ந்த இலக்கிய விருது ஞானபீட விருது.
இந்தியாவில் ராஜஸ்தான் அதிகமாக செம்பு கிடைக்கிறது.
அதிக மக்கள் அடர்த்தியை கொண்ட கண்டம் ஐரோப்பா.
தேசிய காளான் ஆராய்ச்சி மற்றும் பயிற்சி நிலையம் அமைந்துள்ள இடம் சோலன்.உலக அளவில் அதிகமாக பேசப்படும் முதல் மொழி மாண்டரின்.
##
உலக அளவில் அதிகமாக பேசப்படும் முதல் மொழி எது?
##
மாண்டரின்
***
உலகிலேயே மிகப் பெரிய நூலகம் வாஷிங்டன் நூலகம்.இங்கிலாந்திடம் இருந்து அமெரிக்கா 1783 ஆண்டு சுகந்திரம் பெற்றது.“பஞ்சாப் கேசரி ” என்றழைக்கப்பட்ட தேசிய தலைவர் லாலா லஜபதிராய்.சர்க்கரை உற்பத்தியில் முதலாவதாக உள்ள மாநிலம் உத்திரப்பிரதேசம்.புற்று நோய் பற்றி அறியும் பிரிவின் பெயர் ஆங்காலஜி.
##
உலகிலேயே மிகப் பெரிய நூலகம் எங்கு உள்ளது?
##
வாஷிங்டன்
***
உலகிலேயே மிகப் பெரிய நூலகம் வாஷிங்டன் நூலகம்.இங்கிலாந்திடம் இருந்து அமெரிக்கா 1783 ஆண்டு சுகந்திரம் பெற்றது.“பஞ்சாப் கேசரி ” என்றழைக்கப்பட்ட தேசிய தலைவர் லாலா லஜபதிராய்.சர்க்கரை உற்பத்தியில் முதலாவதாக உள்ள மாநிலம் உத்திரப்பிரதேசம்.புற்று நோய் பற்றி அறியும் பிரிவின் பெயர் ஆங்காலஜி.
##
இங்கிலாந்திடம் இருந்து அமெரிக்கா எந்த ஆண்டு சுகந்திரம் பெற்றது?
##
1783
***
உலகிலேயே மிகப் பெரிய நூலகம் வாஷிங்டன் நூலகம்.இங்கிலாந்திடம் இருந்து அமெரிக்கா 1783 ஆண்டு சுகந்திரம் பெற்றது.“பஞ்சாப் கேசரி ” என்றழைக்கப்பட்ட தேசிய தலைவர் லாலா லஜபதிராய்.சர்க்கரை உற்பத்தியில் முதலாவதாக உள்ள மாநிலம் உத்திரப்பிரதேசம்.புற்று நோய் பற்றி அறியும் பிரிவின் பெயர் ஆங்காலஜி.
##
“பஞ்சாப் கேசரி ” என்றழைக்கப்பட்ட தேசிய தலைவர்?
##
லாலா லஜபதிராய்
***
உலகிலேயே மிகப் பெரிய நூலகம் வாஷிங்டன் நூலகம்.இங்கிலாந்திடம் இருந்து அமெரிக்கா 1783 ஆண்டு சுகந்திரம் பெற்றது.“பஞ்சாப் கேசரி ” என்றழைக்கப்பட்ட தேசிய தலைவர் லாலா லஜபதிராய்.சர்க்கரை உற்பத்தியில் முதலாவதாக உள்ள மாநிலம் உத்திரப்பிரதேசம்.புற்று நோய் பற்றி அறியும் பிரிவின் பெயர் ஆங்காலஜி.
##
சர்க்கரை உற்பத்தியில் முதலாவதாக உள்ள மாநிலம்?
##
உத்திரப்பிரதேசம்
***
உலகிலேயே மிகப் பெரிய நூலகம் வாஷிங்டன் நூலகம்.இங்கிலாந்திடம் இருந்து அமெரிக்கா 1783 ஆண்டு சுகந்திரம் பெற்றது.“பஞ்சாப் கேசரி ” என்றழைக்கப்பட்ட தேசிய தலைவர் லாலா லஜபதிராய்.சர்க்கரை உற்பத்தியில் முதலாவதாக உள்ள மாநிலம் உத்திரப்பிரதேசம்.புற்று நோய் பற்றி அறியும் பிரிவின் பெயர் ஆங்காலஜி.
##
புற்று நோய் பற்றி அறியும் பிரிவின் பெயர்?
##
ஆங்காலஜி
***
இந்தியாவின் இரண்டாவது விண்வெளிக் கோள் ரோஹிணி.
இந்தியாவின் முதல் செயற்கைக் கோள் ஆர்யபாட்ட.
உலகிலேயே மிகப்பெரிய நூலகம் மாஸ்கோ நூலகம்.இந்தியாவின் கண்டம் விட்டு கண்டம் பாயும் ஏவுகணையின் பெயர் பிரிதிவி.மனிதன் முதன் முதலில் பயன்படுத்திய திணை வகை கோதுமை.
##
இந்தியாவின் இரண்டாவது விண்வெளிக் கோள்?
##
ரோஹிணி
***
இந்தியாவின் இரண்டாவது விண்வெளிக் கோள் ரோஹிணி.
இந்தியாவின் முதல் செயற்கைக் கோள் ஆர்யபாட்ட.
உலகிலேயே மிகப்பெரிய நூலகம் மாஸ்கோ நூலகம்.இந்தியாவின் கண்டம் விட்டு கண்டம் பாயும் ஏவுகணையின் பெயர் பிரிதிவி.மனிதன் முதன் முதலில் பயன்படுத்திய திணை வகை கோதுமை.
##
இந்தியாவின் முதல் செயற்கைக் கோள் எது?
##
ஆர்யபாட்ட
***
இந்தியாவின் இரண்டாவது விண்வெளிக் கோள் ரோஹிணி.
இந்தியாவின் முதல் செயற்கைக் கோள் ஆர்யபாட்ட.
உலகிலேயே மிகப்பெரிய நூலகம் மாஸ்கோ நூலகம்.இந்தியாவின் கண்டம் விட்டு கண்டம் பாயும் ஏவுகணையின் பெயர் பிரிதிவி.மனிதன் முதன் முதலில் பயன்படுத்திய திணை வகை கோதுமை.
##
உலகிலேயே மிகப்பெரிய நூலகம் எங்கு உள்ளது?
##
மாஸ்கோ
***
இந்தியாவின் இரண்டாவது விண்வெளிக் கோள் ரோஹிணி.
இந்தியாவின் முதல் செயற்கைக் கோள் ஆர்யபாட்ட.
உலகிலேயே மிகப்பெரிய நூலகம் மாஸ்கோ நூலகம்.இந்தியாவின் கண்டம் விட்டு கண்டம் பாயும் ஏவுகணையின் பெயர் பிரிதிவி.மனிதன் முதன் முதலில் பயன்படுத்திய திணை வகை கோதுமை.
##
இந்தியாவின் கண்டம் விட்டு கண்டம் பாயும் ஏவுகணையின் பெயர் என்ன?
##
பிரிதிவி
***
இந்தியாவின் இரண்டாவது விண்வெளிக் கோள் ரோஹிணி.
இந்தியாவின் முதல் செயற்கைக் கோள் ஆர்யபாட்ட.
உலகிலேயே மிகப்பெரிய நூலகம் மாஸ்கோ நூலகம்.இந்தியாவின் கண்டம் விட்டு கண்டம் பாயும் ஏவுகணையின் பெயர் பிரிதிவி.மனிதன் முதன் முதலில் பயன்படுத்திய திணை வகை கோதுமை.
##
மனிதன் முதன் முதலில் பயன்படுத்திய திணை வகை எது?
##
கோதுமை
***
தொழிற்புரட்சி முதன்முதல் ஆரம்பித்த நாடு இங்கிலாந்து.ஜாலியன் வாலாபாக் படுகொலை பஞ்சாப் மாநிலத்தில் நடந்தது.காவிரி நதி வங்காள விரிகுடாவில் பூம்புகார் கலக்கிறது.தென் இந்தியாவின் மான்செஸ்டர் எனப்படுவது கோயமுத்தூர்.“தங்கப் போர்வை நாடு” எனப்படுவது ஆஸ்திரேலியா.
##
தொழிற்புரட்சி முதன்முதல் ஆரம்பித்த நாடு எது?
##
இங்கிலாந்து
***
தொழிற்புரட்சி முதன்முதல் ஆரம்பித்த நாடு இங்கிலாந்து.ஜாலியன் வாலாபாக் படுகொலை பஞ்சாப் மாநிலத்தில் நடந்தது.காவிரி நதி வங்காள விரிகுடாவில் பூம்புகார் கலக்கிறது.தென் இந்தியாவின் மான்செஸ்டர் எனப்படுவது கோயமுத்தூர்.“தங்கப் போர்வை நாடு” எனப்படுவது ஆஸ்திரேலியா.
##
ஜாலியன் வாலாபாக் படுகொலை எந்த மாநிலத்தில் நடந்தது?
##
பஞ்சாப்
***
தொழிற்புரட்சி முதன்முதல் ஆரம்பித்த நாடு இங்கிலாந்து.ஜாலியன் வாலாபாக் படுகொலை பஞ்சாப் மாநிலத்தில் நடந்தது.காவிரி நதி வங்காள விரிகுடாவில் பூம்புகார் கலக்கிறது.தென் இந்தியாவின் மான்செஸ்டர் எனப்படுவது கோயமுத்தூர்.“தங்கப் போர்வை நாடு” எனப்படுவது ஆஸ்திரேலியா.
##
காவிரி நதி வங்காள விரிகுடாவில் எங்கு கலக்கிறது?
##
பூம்புகார்
***
தொழிற்புரட்சி முதன்முதல் ஆரம்பித்த நாடு இங்கிலாந்து.ஜாலியன் வாலாபாக் படுகொலை பஞ்சாப் மாநிலத்தில் நடந்தது.காவிரி நதி வங்காள விரிகுடாவில் பூம்புகார் கலக்கிறது.தென் இந்தியாவின் மான்செஸ்டர் எனப்படுவது கோயமுத்தூர்.“தங்கப் போர்வை நாடு” எனப்படுவது ஆஸ்திரேலியா.
##
“தங்கப் போர்வை நாடு” எனப்படுவது?
##
ஆஸ்திரேலியா
***
தொழிற்புரட்சி முதன்முதல் ஆரம்பித்த நாடு இங்கிலாந்து.ஜாலியன் வாலாபாக் படுகொலை பஞ்சாப் மாநிலத்தில் நடந்தது.காவிரி நதி வங்காள விரிகுடாவில் பூம்புகார் கலக்கிறது.தென் இந்தியாவின் மான்செஸ்டர் எனப்படுவது கோயமுத்தூர்.“தங்கப் போர்வை நாடு” எனப்படுவது ஆஸ்திரேலியா.
##
தென் இந்தியாவின் மான்செஸ்டர் எனப்படுவது?
##
கோயமுத்தூர்
***
நம் நாட்டில் முதன்முதலாக இரும்பு எஃகு தொழிற்சாலை நிறுவப்பட்ட இடம் ஜாம்ஷெட்பூர். ஆப்பச் சோடாவின் வேதியியல் பெயர் சோடியம் பை கார்பனேட்.அணு இணைவு நிகழ்வில் ஏற்படும் ஆற்றல் வெப்ப உட்கரு ஆற்றல்.சருமத்தின் மீதுள்ள நிறத்தின் காரணம் மெலானின்.மிக உயரமான மலைச் சிகரம் நேபாளம் உள்ளது.
##
நம் நாட்டில் முதன்முதலாக இரும்பு எஃகு தொழிற்சாலை நிறுவப்பட்ட இடம் எது?
##
ஜாம்ஷெட்பூர்
***
நம் நாட்டில் முதன்முதலாக இரும்பு எஃகு தொழிற்சாலை நிறுவப்பட்ட இடம் ஜாம்ஷெட்பூர். ஆப்பச் சோடாவின் வேதியியல் பெயர் சோடியம் பை கார்பனேட்.அணு இணைவு நிகழ்வில் ஏற்படும் ஆற்றல் வெப்ப உட்கரு ஆற்றல்.சருமத்தின் மீதுள்ள நிறத்தின் காரணம் மெலானின்.மிக உயரமான மலைச் சிகரம் நேபாளம் உள்ளது.
##
ஆப்பச் சோடாவின் வேதியியல் பெயர்?
##
சோடியம் பை கார்பனேட்
***
நம் நாட்டில் முதன்முதலாக இரும்பு எஃகு தொழிற்சாலை நிறுவப்பட்ட இடம் ஜாம்ஷெட்பூர். ஆப்பச் சோடாவின் வேதியியல் பெயர் சோடியம் பை கார்பனேட்.அணு இணைவு நிகழ்வில் ஏற்படும் ஆற்றல் வெப்ப உட்கரு ஆற்றல்.சருமத்தின் மீதுள்ள நிறத்தின் காரணம் மெலானின்.மிக உயரமான மலைச் சிகரம் நேபாளம் உள்ளது.
##
அணு இணைவு நிகழ்வில் ஏற்படும் ஆற்றல்
##
வெப்ப உட்கரு ஆற்றல்
***
நம் நாட்டில் முதன்முதலாக இரும்பு எஃகு தொழிற்சாலை நிறுவப்பட்ட இடம் ஜாம்ஷெட்பூர். ஆப்பச் சோடாவின் வேதியியல் பெயர் சோடியம் பை கார்பனேட்.அணு இணைவு நிகழ்வில் ஏற்படும் ஆற்றல் வெப்ப உட்கரு ஆற்றல்.சருமத்தின் மீதுள்ள நிறத்தின் காரணம் மெலானின்.மிக உயரமான மலைச் சிகரம் நேபாளம் உள்ளது.
##
சருமத்தின் மீதுள்ள நிறத்தின் காரணம்?
##
மெலானின்
***
நம் நாட்டில் முதன்முதலாக இரும்பு எஃகு தொழிற்சாலை நிறுவப்பட்ட இடம் ஜாம்ஷெட்பூர். ஆப்பச் சோடாவின் வேதியியல் பெயர் சோடியம் பை கார்பனேட்.அணு இணைவு நிகழ்வில் ஏற்படும் ஆற்றல் வெப்ப உட்கரு ஆற்றல்.சருமத்தின் மீதுள்ள நிறத்தின் காரணம் மெலானின்.மிக உயரமான மலைச் சிகரம் நேபாளம் உள்ளது.
##
மிக உயரமான மலைச் சிகரம் எங்கு உள்ளது?
##
நேபாளம்
***
இரண்டாம் உலகப் போர் முடிவடைந்த ஆண்டு 1945.இந்தியாவில் அதிகமாக பேசப்படும் மொழிகளில் இரண்டாவது இடத்தை வகிப்பது தெலுங்கு.	உலகிலேயே மிகப் பெரிய கடற்படை ரஷ்யக் கடற்படை
.ராபிஸ் நோய் உண்டாவதற்குக் காரணம் நாய்க்கடி.
டயா காந்தப் பொருளுக்கு எடுத்துக் காட்டு பாதரசம்.
##
இரண்டாம் உலகப் போர் முடிவடைந்த ஆண்டு எது?
##
1945
***
இரண்டாம் உலகப் போர் முடிவடைந்த ஆண்டு 1945.இந்தியாவில் அதிகமாக பேசப்படும் மொழிகளில் இரண்டாவது இடத்தை வகிப்பது தெலுங்கு.	உலகிலேயே மிகப் பெரிய கடற்படை ரஷ்யக் கடற்படை
.ராபிஸ் நோய் உண்டாவதற்குக் காரணம் நாய்க்கடி.
டயா காந்தப் பொருளுக்கு எடுத்துக் காட்டு பாதரசம்.
##
இந்தியாவில் அதிகமாக பேசப்படும் மொழிகளில் இரண்டாவது இடத்தை வகிப்பது எது?
##
தெலுங்கு
***
இரண்டாம் உலகப் போர் முடிவடைந்த ஆண்டு 1945.இந்தியாவில் அதிகமாக பேசப்படும் மொழிகளில் இரண்டாவது இடத்தை வகிப்பது தெலுங்கு.	உலகிலேயே மிகப் பெரிய கடற்படை ரஷ்யக் கடற்படை
.ராபிஸ் நோய் உண்டாவதற்குக் காரணம் நாய்க்கடி.
டயா காந்தப் பொருளுக்கு எடுத்துக் காட்டு பாதரசம்.
##
உலகிலேயே மிகப் பெரிய கடற்படை எது?
##
ரஷ்யக் கடற்படை
***
இரண்டாம் உலகப் போர் முடிவடைந்த ஆண்டு 1945.இந்தியாவில் அதிகமாக பேசப்படும் மொழிகளில் இரண்டாவது இடத்தை வகிப்பது தெலுங்கு.	உலகிலேயே மிகப் பெரிய கடற்படை ரஷ்யக் கடற்படை
.ராபிஸ் நோய் உண்டாவதற்குக் காரணம் நாய்க்கடி.
டயா காந்தப் பொருளுக்கு எடுத்துக் காட்டு பாதரசம்.
##
ராபிஸ் நோய் உண்டாவதற்குக் காரணம் எது?
##
நாய்க்கடி
***
இரண்டாம் உலகப் போர் முடிவடைந்த ஆண்டு 1945.இந்தியாவில் அதிகமாக பேசப்படும் மொழிகளில் இரண்டாவது இடத்தை வகிப்பது தெலுங்கு.	உலகிலேயே மிகப் பெரிய கடற்படை ரஷ்யக் கடற்படை
.ராபிஸ் நோய் உண்டாவதற்குக் காரணம் நாய்க்கடி.
டயா காந்தப் பொருளுக்கு எடுத்துக் காட்டு பாதரசம்
##
டயா காந்தப் பொருளுக்கு எடுத்துக் காட்டு எது?
##
பாதரசம்
***
குறிஞ்சி மலர் என்ற சிறுகதையின் ஆசிரியர் நா.பார்த்தசாரதி.பூமிதான இயக்கத்தை தொடங்கியவர்  ஆச்சார் வினோபாபாவே.தமிழ்நாட்டில் கரும்பு ஆராய்ச்சி நிலையம் சிறுகமணி.	புத்த சரித்திரத்தின் ஆசிரியர் அஸ்வகோஷா.மழைத்துளிகள் கோள வடிவத்தை பெறக்காரணம் பரப்பு இழுவிசை.
##
குறிஞ்சி மலர் என்ற சிறுகதையின் ஆசிரியர் யார்?
##
நா.பார்த்தசாரதி
***
குறிஞ்சி மலர் என்ற சிறுகதையின் ஆசிரியர் நா.பார்த்தசாரதி.பூமிதான இயக்கத்தை தொடங்கியவர்  ஆச்சார் வினோபாபாவே.தமிழ்நாட்டில் கரும்பு ஆராய்ச்சி நிலையம் சிறுகமணி.	புத்த சரித்திரத்தின் ஆசிரியர் அஸ்வகோஷா.மழைத்துளிகள் கோள வடிவத்தை பெறக்காரணம் பரப்பு இழுவிசை.
##
பூமிதான இயக்கத்தை தொடங்கியவர் யார்?
##
வினோபாபாவே
***
குறிஞ்சி மலர் என்ற சிறுகதையின் ஆசிரியர் நா.பார்த்தசாரதி.பூமிதான இயக்கத்தை தொடங்கியவர்  ஆச்சார் வினோபாபாவே.தமிழ்நாட்டில் கரும்பு ஆராய்ச்சி நிலையம் சிறுகமணி.	புத்த சரித்திரத்தின் ஆசிரியர் அஸ்வகோஷா.மழைத்துளிகள் கோள வடிவத்தை பெறக்காரணம் பரப்பு இழுவிசை.
##
தமிழ்நாட்டில் கரும்பு ஆராய்ச்சி நிலையம் எங்குள்ளது?
##
சிறுகமணி
***
குறிஞ்சி மலர் என்ற சிறுகதையின் ஆசிரியர் நா.பார்த்தசாரதி.பூமிதான இயக்கத்தை தொடங்கியவர்  ஆச்சார் வினோபாபாவே.தமிழ்நாட்டில் கரும்பு ஆராய்ச்சி நிலையம் சிறுகமணி.	புத்த சரித்திரத்தின் ஆசிரியர் அஸ்வகோஷா.மழைத்துளிகள் கோள வடிவத்தை பெறக்காரணம் பரப்பு இழுவிசை.
##
புத்த சரித்திரத்தின் ஆசிரியர் யார்?
##
அஸ்வகோஷா
***
குறிஞ்சி மலர் என்ற சிறுகதையின் ஆசிரியர் நா.பார்த்தசாரதி.பூமிதான இயக்கத்தை தொடங்கியவர்  ஆச்சார் வினோபாபாவே.தமிழ்நாட்டில் கரும்பு ஆராய்ச்சி நிலையம் சிறுகமணி.	புத்த சரித்திரத்தின் ஆசிரியர் அஸ்வகோஷா.மழைத்துளிகள் கோள வடிவத்தை பெறக்காரணம் பரப்பு இழுவிசை.
##
மழைத்துளிகள் கோள வடிவத்தை பெறக்காரணம் எது?
##
பரப்பு இழுவிசை
***
ஹீமோகுளோபினில் உள்ள உலோகம் இரும்பு.புல்லி என்ற வார்த்தை தொடர்பு கொண்டது ஹாக்கி.இந்தியாவின் தேசிய மரம் ஆலமரம்.பூட்டான் நாட்டின் தலைநகர் திம்பு.இன்சுலினை சுரக்கும் உறுப்பு கணையம்.
##
ஹீமோகுளோபினில் உள்ள உலோகம் எது?
##
இரும்பு
***
ஹீமோகுளோபினில் உள்ள உலோகம் இரும்பு.புல்லி என்ற வார்த்தை தொடர்பு கொண்டது ஹாக்கி.இந்தியாவின் தேசிய மரம் ஆலமரம்.பூட்டான் நாட்டின் தலைநகர் திம்பு.இன்சுலினை சுரக்கும் உறுப்பு கணையம்.
##
புல்லி என்ற வார்த்தை தொடர்பு கொண்டது எது?
##
ஹாக்கி
***
ஹீமோகுளோபினில் உள்ள உலோகம் இரும்பு.புல்லி என்ற வார்த்தை தொடர்பு கொண்டது ஹாக்கி.இந்தியாவின் தேசிய மரம் ஆலமரம்.பூட்டான் நாட்டின் தலைநகர் திம்பு.இன்சுலினை சுரக்கும் உறுப்பு கணையம்.
##
இந்தியாவின் தேசிய மரம் எது?
##
ஆலமரம்
***
ஹீமோகுளோபினில் உள்ள உலோகம் இரும்பு.புல்லி என்ற வார்த்தை தொடர்பு கொண்டது ஹாக்கி.இந்தியாவின் தேசிய மரம் ஆலமரம்.பூட்டான் நாட்டின் தலைநகர் திம்பு.இன்சுலினை சுரக்கும் உறுப்பு கணையம்.
##
பூட்டான் நாட்டின் தலைநகர் எது?
##
திம்பு
***
ஹீமோகுளோபினில் உள்ள உலோகம் இரும்பு.புல்லி என்ற வார்த்தை தொடர்பு கொண்டது ஹாக்கி.இந்தியாவின் தேசிய மரம் ஆலமரம்.பூட்டான் நாட்டின் தலைநகர் திம்பு.இன்சுலினை சுரக்கும் உறுப்பு கணையம்.
##
இன்சுலினை சுரக்கும் உறுப்பு எது?
##
கணையம்
***
உலகப் புகழ்பெற்ற சாய்ந்த கோபுரம் இத்தாலி உள்ளது.இந்தியாவில் அதிக மலைவாழ் மக்கள் வாழும் மாநிலம் உத்திரப் பிரதேசம்.சுங்க வம்சத்தை நிறுவியவர் புஷ்யமித்திரர்.மிகப்பெரிய தரைகடல் மத்திய தரைகடல்
.பாக்சைட்டில் கிடைப்பது அலுமினியம்.
##
உலகப் புகழ்பெற்ற சாய்ந்த கோபுரம் எங்கு உள்ளது?
##
இத்தாலி
***
உலகப் புகழ்பெற்ற சாய்ந்த கோபுரம் இத்தாலி உள்ளது.இந்தியாவில் அதிக மலைவாழ் மக்கள் வாழும் மாநிலம் உத்திரப் பிரதேசம்.சுங்க வம்சத்தை நிறுவியவர் புஷ்யமித்திரர்.மிகப்பெரிய தரைகடல் மத்திய தரைகடல்
.பாக்சைட்டில் கிடைப்பது அலுமினியம்.
##
இந்தியாவில் அதிக மலைவாழ் மக்கள் வாழும் மாநிலம் எது?
##
உத்திரப் பிரதேசம்
***
உலகப் புகழ்பெற்ற சாய்ந்த கோபுரம் இத்தாலி உள்ளது.இந்தியாவில் அதிக மலைவாழ் மக்கள் வாழும் மாநிலம் உத்திரப் பிரதேசம்.சுங்க வம்சத்தை நிறுவியவர் புஷ்யமித்திரர்.மிகப்பெரிய தரைகடல் மத்திய தரைகடல்
.பாக்சைட்டில் கிடைப்பது அலுமினியம்.
##
சுங்க வம்சத்தை நிறுவியவர் யார்?
##
புஷ்யமித்திரர்
***
உலகப் புகழ்பெற்ற சாய்ந்த கோபுரம் இத்தாலி உள்ளது.இந்தியாவில் அதிக மலைவாழ் மக்கள் வாழும் மாநிலம் உத்திரப் பிரதேசம்.சுங்க வம்சத்தை நிறுவியவர் புஷ்யமித்திரர்.மிகப்பெரிய தரைகடல் மத்திய தரைகடல்
.பாக்சைட்டில் கிடைப்பது அலுமினியம்.
##
மிகப்பெரிய தரைகடல் எது?
##
மத்திய தரைகடல்
***
உலகப் புகழ்பெற்ற சாய்ந்த கோபுரம் இத்தாலி உள்ளது.இந்தியாவில் அதிக மலைவாழ் மக்கள் வாழும் மாநிலம் உத்திரப் பிரதேசம்.சுங்க வம்சத்தை நிறுவியவர் புஷ்யமித்திரர்.மிகப்பெரிய தரைகடல் மத்திய தரைகடல்
.பாக்சைட்டில் கிடைப்பது அலுமினியம்.
##
பாக்சைட்டில் கிடைப்பது?
##
அலுமினியம்
***
போலியோ தடுப்பூசி மருந்தைக் கண்டுபிடித்தவர் ஜோனார்ஸ் சால்க்.பூஜ்யத்தைக் கண்டுபிடித்தவர் இந்தியர் நாட்டினர். விதையில்லாத தாவரங்கள்
கிரிப்டோகேம்.ஆக்சிஜன் படகு ஹீமோகுளோபின்.
திரு.வி.க. நடத்தி வந்த பத்திரிகை தேசபக்தன்.
##
போலியோ தடுப்பூசி மருந்தைக் கண்டுபிடித்தவர் யார்?
##
ஜோனார்ஸ் சால்க்
***
போலியோ தடுப்பூசி மருந்தைக் கண்டுபிடித்தவர் ஜோனார்ஸ் சால்க்.பூஜ்யத்தைக் கண்டுபிடித்தவர் இந்தியர் நாட்டினர். விதையில்லாத தாவரங்கள்
கிரிப்டோகேம்.ஆக்சிஜன் படகு ஹீமோகுளோபின்.
திரு.வி.க. நடத்தி வந்த பத்திரிகை தேசபக்தன்.
##
பூஜ்யத்தைக் கண்டுபிடித்தவர் எந்த நாட்டினர்?
##
இந்தியர்
***
போலியோ தடுப்பூசி மருந்தைக் கண்டுபிடித்தவர் ஜோனார்ஸ் சால்க்.பூஜ்யத்தைக் கண்டுபிடித்தவர் இந்தியர் நாட்டினர். விதையில்லாத தாவரங்கள்
கிரிப்டோகேம்.ஆக்சிஜன் படகு ஹீமோகுளோபின்.
திரு.வி.க. நடத்தி வந்த பத்திரிகை தேசபக்தன்.
##
விதையில்லாத தாவரங்கள்
##
கிரிப்டோகேம்
***
போலியோ தடுப்பூசி மருந்தைக் கண்டுபிடித்தவர் ஜோனார்ஸ் சால்க்.பூஜ்யத்தைக் கண்டுபிடித்தவர் இந்தியர் நாட்டினர். விதையில்லாத தாவரங்கள்
கிரிப்டோகேம்.ஆக்சிஜன் படகு ஹீமோகுளோபின்.
திரு.வி.க. நடத்தி வந்த பத்திரிகை தேசபக்தன்.
##
ஆக்சிஜன் படகு என்று கூறுவது?
##
ஹீமோகுளோபின்
***
போலியோ தடுப்பூசி மருந்தைக் கண்டுபிடித்தவர் ஜோனார்ஸ் சால்க்.பூஜ்யத்தைக் கண்டுபிடித்தவர் இந்தியர் நாட்டினர். விதையில்லாத தாவரங்கள்
கிரிப்டோகேம்.ஆக்சிஜன் படகு ஹீமோகுளோபின்.
திரு.வி.க. நடத்தி வந்த பத்திரிகை தேசபக்தன்.
##
திரு.வி.க. நடத்தி வந்த பத்திரிகை எது?
##
தேசபக்தன்
***
இந்தியாவில் இஞ்சி அதிகமாக விளையும் மாநிலம் கர்நாடகம்.இந்தியாவில் மேற்கத்திய கல்வி முறைக்கு வித்திட்டாவர் மெக்காலே.வருமான வரி என்பது  ஒரு நேர்முக வரி.இந்தியாவின் முதல் குடிமகன் குடியரசுத் தலைவர்.இளைஞர் தினம் தொடர்புடையது விவேகானந்தர்.
##
இந்தியாவில் இஞ்சி அதிகமாக விளையும் மாநிலம் எது?
##
கர்நாடகம்
***
இந்தியாவில் இஞ்சி அதிகமாக விளையும் மாநிலம் கர்நாடகம்.இந்தியாவில் மேற்கத்திய கல்வி முறைக்கு வித்திட்டாவர் மெக்காலே.வருமான வரி என்பது  ஒரு நேர்முக வரி.இந்தியாவின் முதல் குடிமகன் குடியரசுத் தலைவர்.இளைஞர் தினம் தொடர்புடையது விவேகானந்தர்.
##
இந்தியாவில் மேற்கத்திய கல்வி முறைக்கு வித்திட்டாவர் யார்?
##
மெக்காலே
***
இந்தியாவில் இஞ்சி அதிகமாக விளையும் மாநிலம் கர்நாடகம்.இந்தியாவில் மேற்கத்திய கல்வி முறைக்கு வித்திட்டாவர் மெக்காலே.வருமான வரி என்பது  ஒரு நேர்முக வரி.இந்தியாவின் முதல் குடிமகன் குடியரசுத் தலைவர்.இளைஞர் தினம் தொடர்புடையது விவேகானந்தர்.
##
வருமான வரி என்பது?
##
ஒரு மறைமுக வரி
***
இந்தியாவில் இஞ்சி அதிகமாக விளையும் மாநிலம் கர்நாடகம்.இந்தியாவில் மேற்கத்திய கல்வி முறைக்கு வித்திட்டாவர் மெக்காலே.வருமான வரி என்பது  ஒரு நேர்முக வரி.இந்தியாவின் முதல் குடிமகன் குடியரசுத் தலைவர்.இளைஞர் தினம் தொடர்புடையது விவேகானந்தர்.
##
இந்தியாவின் முதல் குடிமகன் யார்?
##
குடியரசுத் தலைவர்
***
இந்தியாவில் இஞ்சி அதிகமாக விளையும் மாநிலம் கர்நாடகம்.இந்தியாவில் மேற்கத்திய கல்வி முறைக்கு வித்திட்டாவர் மெக்காலே.வருமான வரி என்பது  ஒரு நேர்முக வரி.இந்தியாவின் முதல் குடிமகன் குடியரசுத் தலைவர்.இளைஞர் தினம் தொடர்புடையது விவேகானந்தர்.
##
இளைஞர் தினம் தொடர்புடையது
##
விவேகானந்தர்
***
அமெரிக்க ஐக்கிய நாடுகள் மற்றும் இங்கிலாந்து சீனாவிற்காக ஏற்படுத்திய கொள்கை திறந்த வெளிக்கொள்கை.துருக்கி தன் ஆதரவை வழங்கியது
மைய நாடுகள்.பாசிச கட்சியை தோற்றுவித்தவர்
லெனின்.ஜெர்மனி மற்றும் இத்தாலி ஒன்றிணைக்கப்பட்ட ஆண்டு 1870.ஆங்கில கிழக்கிந்திய வாணிகக் குழு தோற்றுவிக்கப்பட்ட ஆண்டு  கி.பி. 1660.
##
அமெரிக்க ஐக்கிய நாடுகள் மற்றும் இங்கிலாந்து சீனாவிற்காக ஏற்படுத்திய கொள்கை
##
திறந்த வெளிக்கொள்கை
***
அமெரிக்க ஐக்கிய நாடுகள் மற்றும் இங்கிலாந்து சீனாவிற்காக ஏற்படுத்திய கொள்கை திறந்த வெளிக்கொள்கை.துருக்கி தன் ஆதரவை வழங்கியது
மைய நாடுகள்.பாசிச கட்சியை தோற்றுவித்தவர்
லெனின்.ஜெர்மனி மற்றும் இத்தாலி ஒன்றிணைக்கப்பட்ட ஆண்டு 1870.ஆங்கில கிழக்கிந்திய வாணிகக் குழு தோற்றுவிக்கப்பட்ட ஆண்டு  கி.பி. 1660.
##
துருக்கி தன் ஆதரவை வழங்கியது?
##
மைய நாடுகள்
***
அமெரிக்க ஐக்கிய நாடுகள் மற்றும் இங்கிலாந்து சீனாவிற்காக ஏற்படுத்திய கொள்கை திறந்த வெளிக்கொள்கை.துருக்கி தன் ஆதரவை வழங்கியது
மைய நாடுகள்.பாசிச கட்சியை தோற்றுவித்தவர்
லெனின்.ஜெர்மனி மற்றும் இத்தாலி ஒன்றிணைக்கப்பட்ட ஆண்டு 1870.ஆங்கில கிழக்கிந்திய வாணிகக் குழு தோற்றுவிக்கப்பட்ட ஆண்டு  கி.பி. 1660.
##
பாசிச கட்சியை தோற்றுவித்தவர்
##
லெனின்
***
அமெரிக்க ஐக்கிய நாடுகள் மற்றும் இங்கிலாந்து சீனாவிற்காக ஏற்படுத்திய கொள்கை திறந்த வெளிக்கொள்கை.துருக்கி தன் ஆதரவை வழங்கியது
மைய நாடுகள்.பாசிச கட்சியை தோற்றுவித்தவர்
லெனின்.ஜெர்மனி மற்றும் இத்தாலி ஒன்றிணைக்கப்பட்ட ஆண்டு 1870.ஆங்கில கிழக்கிந்திய வாணிகக் குழு தோற்றுவிக்கப்பட்ட ஆண்டு  கி.பி. 1660.
##
ஜெர்மனி மற்றும் இத்தாலி ஒன்றிணைக்கப்பட்ட ஆண்டு?
##
1870
***
அமெரிக்க ஐக்கிய நாடுகள் மற்றும் இங்கிலாந்து சீனாவிற்காக ஏற்படுத்திய கொள்கை திறந்த வெளிக்கொள்கை.துருக்கி தன் ஆதரவை வழங்கியது
மைய நாடுகள்.பாசிச கட்சியை தோற்றுவித்தவர்
லெனின்.ஜெர்மனி மற்றும் இத்தாலி ஒன்றிணைக்கப்பட்ட ஆண்டு 1870.ஆங்கில கிழக்கிந்திய வாணிகக் குழு தோற்றுவிக்கப்பட்ட ஆண்டு  கி.பி. 1660.
##
ஆங்கில கிழக்கிந்திய வாணிகக் குழு தோற்றுவிக்கப்பட்ட ஆண்டு?
##
கி.பி. 1660
***
ஐரோப்பிய நாடுகள் 'செல்வாக்கை நிலைநாட்டுதல்' என்ற கொள்கையைப் பின்பற்றியது சீனா.பொருள்களின் போக்குவரத்தை அதிகரிக்கச் செய்தது இரயில்வே.சீனக் குடியரசை உருவாக்கியவர் டாக்டர் சன்யாட் சென்.ஜெர்மனியால் மூழ்கடிக்கப்பட்ட அமெரிக்காவின் புகழ் பெற்ற வணிகக் கப்பல் லூசிடானியா. முதல் உலகப்போருக்குப்பின் வல்லரசாக எழுச்சி பெற்ற நாடு ஜப்பான்.
##
ஐரோப்பிய நாடுகள் 'செல்வாக்கை நிலைநாட்டுதல்' என்ற கொள்கையைப் பின்பற்றியது?
##
சீனா
***
ஐரோப்பிய நாடுகள் 'செல்வாக்கை நிலைநாட்டுதல்' என்ற கொள்கையைப் பின்பற்றியது சீனா.பொருள்களின் போக்குவரத்தை அதிகரிக்கச் செய்தது இரயில்வே.சீனக் குடியரசை உருவாக்கியவர் டாக்டர் சன்யாட் சென்.ஜெர்மனியால் மூழ்கடிக்கப்பட்ட அமெரிக்காவின் புகழ் பெற்ற வணிகக் கப்பல் லூசிடானியா. முதல் உலகப்போருக்குப்பின் வல்லரசாக எழுச்சி பெற்ற நாடு ஜப்பான்.
##
பொருள்களின் போக்குவரத்தை அதிகரிக்கச் செய்தது?
##
இரயில்வே
***
ஐரோப்பிய நாடுகள் 'செல்வாக்கை நிலைநாட்டுதல்' என்ற கொள்கையைப் பின்பற்றியது சீனா.பொருள்களின் போக்குவரத்தை அதிகரிக்கச் செய்தது இரயில்வே.சீனக் குடியரசை உருவாக்கியவர் டாக்டர் சன்யாட் சென்.ஜெர்மனியால் மூழ்கடிக்கப்பட்ட அமெரிக்காவின் புகழ் பெற்ற வணிகக் கப்பல் லூசிடானியா. முதல் உலகப்போருக்குப்பின் வல்லரசாக எழுச்சி பெற்ற நாடு ஜப்பான்.
##
சீனக் குடியரசை உருவாக்கியவர்?
##
டாக்டர் சன்யாட் சென்
***
ஐரோப்பிய நாடுகள் 'செல்வாக்கை நிலைநாட்டுதல்' என்ற கொள்கையைப் பின்பற்றியது சீனா.பொருள்களின் போக்குவரத்தை அதிகரிக்கச் செய்தது இரயில்வே.சீனக் குடியரசை உருவாக்கியவர் டாக்டர் சன்யாட் சென்.ஜெர்மனியால் மூழ்கடிக்கப்பட்ட அமெரிக்காவின் புகழ் பெற்ற வணிகக் கப்பல் லூசிடானியா. முதல் உலகப்போருக்குப்பின் வல்லரசாக எழுச்சி பெற்ற நாடு ஜப்பான்.
##
ஜெர்மனியால் மூழ்கடிக்கப்பட்ட அமெரிக்காவின் புகழ் பெற்ற வணிகக் கப்பல்?
##
லூசிடானியா
***
ஐரோப்பிய நாடுகள் 'செல்வாக்கை நிலைநாட்டுதல்' என்ற கொள்கையைப் பின்பற்றியது சீனா.பொருள்களின் போக்குவரத்தை அதிகரிக்கச் செய்தது இரயில்வே.சீனக் குடியரசை உருவாக்கியவர் டாக்டர் சன்யாட் சென்.ஜெர்மனியால் மூழ்கடிக்கப்பட்ட அமெரிக்காவின் புகழ் பெற்ற வணிகக் கப்பல் லூசிடானியா. முதல் உலகப்போருக்குப்பின் வல்லரசாக எழுச்சி பெற்ற நாடு ஜப்பான்.
##
முதல் உலகப்போருக்குப்பின் வல்லரசாக எழுச்சி பெற்ற நாடு?
##
ஜப்பான்
***
ஆங்கிலேய கிழக்கிந்திய குழுமத்திற்கு சூரத்தில் வணிகமையம் அமைக்க அனுமதியளித்த முகலாய மன்னர் ஜஹாங்கீர்.பொருளாதாரப் பெருமந்தம் தோன்றிய நாடு அமெரிக்கா.முதல் உலகப் போரை முடிவுக்குக் கொண்டு வந்தது  பாரிஸ் அமைதி மாநாடு.
பிரெஞ்சு கிழக்கிந்திய வணிகக் குழுவை நிறுவியவர் கால்பர்ட்.ஹிட்லர் வியன்னாவில் பணியாற்றியது பெயிண்டர்.
##
ஆங்கிலேய கிழக்கிந்திய குழுமத்திற்கு சூரத்தில் வணிகமையம் அமைக்க அனுமதியளித்த முகலாய மன்னர் யார் ?
##
ஜஹாங்கீர்
***
ஆங்கிலேய கிழக்கிந்திய குழுமத்திற்கு சூரத்தில் வணிகமையம் அமைக்க அனுமதியளித்த முகலாய மன்னர் ஜஹாங்கீர்.பொருளாதாரப் பெருமந்தம் தோன்றிய நாடு அமெரிக்கா.முதல் உலகப் போரை முடிவுக்குக் கொண்டு வந்தது  பாரிஸ் அமைதி மாநாடு.
பிரெஞ்சு கிழக்கிந்திய வணிகக் குழுவை நிறுவியவர் கால்பர்ட்.ஹிட்லர் வியன்னாவில் பணியாற்றியது பெயிண்டர்.
##
பொருளாதாரப் பெருமந்தம் தோன்றிய நாடு எது? 
##
அமெரிக்கா 
***
ஆங்கிலேய கிழக்கிந்திய குழுமத்திற்கு சூரத்தில் வணிகமையம் அமைக்க அனுமதியளித்த முகலாய மன்னர் ஜஹாங்கீர்.பொருளாதாரப் பெருமந்தம் தோன்றிய நாடு அமெரிக்கா.முதல் உலகப் போரை முடிவுக்குக் கொண்டு வந்தது  பாரிஸ் அமைதி மாநாடு.
பிரெஞ்சு கிழக்கிந்திய வணிகக் குழுவை நிறுவியவர் கால்பர்ட்.ஹிட்லர் வியன்னாவில் பணியாற்றியது பெயிண்டர்.
##
முதல் உலகப் போரை முடிவுக்குக் கொண்டு வந்தது எது?
##
பாரிஸ் அமைதி மாநாடு
***
ஆங்கிலேய கிழக்கிந்திய குழுமத்திற்கு சூரத்தில் வணிகமையம் அமைக்க அனுமதியளித்த முகலாய மன்னர் ஜஹாங்கீர்.பொருளாதாரப் பெருமந்தம் தோன்றிய நாடு அமெரிக்கா.முதல் உலகப் போரை முடிவுக்குக் கொண்டு வந்தது  பாரிஸ் அமைதி மாநாடு.
பிரெஞ்சு கிழக்கிந்திய வணிகக் குழுவை நிறுவியவர் கால்பர்ட்.ஹிட்லர் வியன்னாவில் பணியாற்றியது பெயிண்டர்.
##
பிரெஞ்சு கிழக்கிந்திய வணிகக் குழுவை நிறுவியவர் யார் ?
##
கால்பர்ட்
***
ஆங்கிலேய கிழக்கிந்திய குழுமத்திற்கு சூரத்தில் வணிகமையம் அமைக்க அனுமதியளித்த முகலாய மன்னர் ஜஹாங்கீர்.பொருளாதாரப் பெருமந்தம் தோன்றிய நாடு அமெரிக்கா.முதல் உலகப் போரை முடிவுக்குக் கொண்டு வந்தது  பாரிஸ் அமைதி மாநாடு.
பிரெஞ்சு கிழக்கிந்திய வணிகக் குழுவை நிறுவியவர் கால்பர்ட்.ஹிட்லர் வியன்னாவில் பணியாற்றியது பெயிண்டர்.
##
ஹிட்லர் வியன்னாவில் பணியாற்றியது எது?
##
பெயிண்டர்
***
சோடியத்தின் அருகிலுள்ள மத்த வாயு நியான்.பிரகாசமான ஒளியுடன் எரியும் தனிமம் மக்னீசியம்.கேரட் அதிக அளவு காணப்படுவது வைட்டமின் எ.கோழி குஞ்சு பொறிக்க 21 நாட்கள் காலம் அடைக்காக்கும்.பயோரியா வியாதியால் உடலின் பற்கள் பாதிக்கப்படும்.
##
சோடியத்தின் அருகிலுள்ள மத்த வாயு எது?
##
நியான்
***
சோடியத்தின் அருகிலுள்ள மத்த வாயு நியான்.பிரகாசமான ஒளியுடன் எரியும் தனிமம் மக்னீசியம்.கேரட் அதிக அளவு காணப்படுவதுவைட்டமின் எ.கோழி குஞ்சு பொறிக்க 21 நாட்கள் காலம் அடைக்காக்கும்.பயோரியா வியாதியால் உடலின் பற்கள் பாதிக்கப்படும்.
##
பிரகாசமான ஒளியுடன் எரியும் தனிமம் எது?
##
மக்னீசியம்
***
சோடியத்தின் அருகிலுள்ள மத்த வாயு நியான்.பிரகாசமான ஒளியுடன் எரியும் தனிமம் மக்னீசியம்.கேரட் அதிக அளவு காணப்படுவது வைட்டமின் எ.கோழி குஞ்சு பொறிக்க 21 நாட்கள் காலம் அடைக்காக்கும்.பயோரியா வியாதியால் உடலின் பற்கள் பாதிக்கப்படும்.
##
கேரட் அதிக அளவு காணப்படுவது எது?
##
வைட்டமின் எ
***
சோடியத்தின் அருகிலுள்ள மத்த வாயு நியான்.பிரகாசமான ஒளியுடன் எரியும் தனிமம் மக்னீசியம்.கேரட் அதிக அளவு காணப்படுவதுவைட்டமின் எ.கோழி குஞ்சு பொறிக்க 21 நாட்கள் காலம் அடைக்காக்கும்.பயோரியா வியாதியால் உடலின் பற்கள் பாதிக்கப்படும்.
##
கோழி குஞ்சு பொறிக்க எவ்வளவு காலம் அடைக்காக்கும்?
##
21 நாட்கள்
***
சோடியத்தின் அருகிலுள்ள மத்த வாயு நியான்.பிரகாசமான ஒளியுடன் எரியும் தனிமம் மக்னீசியம்.கேரட் அதிக அளவு காணப்படுவதுவைட்டமின் எ.கோழி குஞ்சு பொறிக்க 21 நாட்கள் காலம் அடைக்காக்கும்.பயோரியா வியாதியால் உடலின் பற்கள் பாதிக்கப்படும்.
##
பயோரியா வியாதியால் உடலின் எந்தப்பகுதி பாதிக்கப்படும்?
##
பற்கள்
***
தமிழ், திராவிட மொழிக் குடும்பத்தைச் சேர்ந்த ஒரு மொழியாகும். இக்குடும்பத்தில், இருளா, கைக்காடி, பெட்டக் குறும்பா, சோலகா மற்றும் யெருகுலா என்னும் மொழிகள் அடங்கும். தமிழ் மொழிக் குடும்பம்.
##
தமிழ் எந்த குடும்மத்தை சேர்ந்த ஒரு மொழியாகும் ?
##
திராவிட மொழிக்

***
அணிகளைக் கொண்ட ஒரு குழுவில் ஒவ்வோர் அணியும் ஏனைய ஒன்பது அணிகளுடன் ஒரு முறை மோதின.தொடர் முழுவதும் சிறப்பாக ஆடிய வீரர்களைக் கொண்டு தொடரின் அணி உருவாக்கப்படுகிறது. அந்த வகையில் 15 சூலை 2019 நாளன்று கேன் வில்லியம்சன் தலைமையிலான தொடரின் அணியை ஐசிசி அறிவித்தது.
##
யார் தலைமையிலான தொடரின் அணியை ஐசிசி அறிவித்தது?
##
கேன் வில்லியம்சன்

***  
பழைய எகிப்து இராச்சியத்தின் வரலாறு கிமு 2686-இல் துவங்கி, கிமு 2181 முடிய விளங்கியது
## 
பழைய எகிப்து இராச்சியத்தின் வரலாறு எப்போது துவங்கியது?
## 
கிமு 2686 இல் துவங்கியது.

***   
புழுப்பாம்பு அனைத்தும் பெண் பாம்புகளாக இருப்பதால், ஆண் துணையின்றி இனப்பெருக்கம் செய்பவையாக உள்ளன.
##
எந்த பாம்பு ஆண் துணையின்றி இனப்பெருக்கம் செய்யும்?
## 
புழுப்பாம்பு ஆண் துணையின்றி இனப்பெருக்கம் செய்யும்.

***  
காந்திக்கு மகாத்மா என்னும் கௌரவத்தை வழங்கியவர் இரவீந்திரநாத் தாகூர் ஆவார்.
## 
காந்திக்கு மகாத்மா என்னும் கௌரவத்தை வழங்கியவர் யார்?
## 
காந்திக்கு மகாத்மா என்னும் கௌரவத்தை வழங்கியவர் இரவீந்திரநாத் தாகூர்.

*** 
காந்தி 2 அக்டோபர் 1869 அன்று இந்திய நாட்டின் குஜராத் மாநிலத்திலுள்ள போர்பந்தர் எனும் ஊரில் பிறந்தார்.
## 
காந்தி இங்கு பிறந்தார்?
## 
காந்தி குஜராத் மாநிலத்திலுள்ள போர்பந்தர் எனும் ஊரில் பிறந்தார்.

*** 
தமிழ்நாட்டின் தலைநகரமாக சென்னை உள்ளது.
##  
தமிழ்நாட்டின் தலைநகரம் எது?
## 
தமிழ்நாட்டின் தலைநகரம் சென்னை.

***  
ஆக்சுபோர்டு பல்கலைக்கழகமானது இங்கிலாந்தில், ஆக்சுபோர்டு என்னும் நகரத்தில் அமைந்துள்ள தொன்மைப் புகழ் வாய்ந்த பல்கலைக்கழகமாகும்.
##  
ஆக்சுபோர்டு பல்கலைக்கழகம் எங்கு உள்ளது?
##  
ஆக்சுபோர்டு பல்கலைக்கழகமானது இங்கிலாந்தில் உள்ளது.

***  
ஜவகர்லால் நேருவின் பிறந்தநாள் அன்று இந்தியாவில் குழந்தைகள் தினம் கொண்டாடப்படுகிறது.
##  
யாருடைய பிறந்தநாள் அன்று இந்தியாவில் குழந்தைகள் தினம் கொண்டாடப்படுகிறது?
##  
ஜவகர்லால் நேருவின்.

*** 
இராணி வேலுநாச்சியார் இந்தியாவின் முதல் பெண் விடுதலைப் போராட்ட வீராங்கனை ஆவார்.
## 
இந்தியாவின் முதல் பெண் விடுதலைப் போராட்ட வீராங்கனை யார்?
##  
இராணி வேலுநாச்சியார்.

***  
பாக்குநீரிணையை முதன் முதலில் நீந்திக் கடந்தவர் என்ற பெருமை பெற்றவர் முருகுப்பிள்ளை நவரத்தினசாமி.
##  
பாக்குநீரிணையை முதன் முதலில் நீந்திக் கடந்தவர் யார்?
##  
முருகுப்பிள்ளை நவரத்தினசாமி

*** 
தடுக்கப்பட்ட நகர் அல்லது பெய்சிங் அரச அரண்மனை அல்லது அரண்மனை அருங்காட்சியகம் எனப்படுவது சீன தலைநகர் பெய்சிங்கின் நடுவில் அமைந்துள்ள, பழம்பெருமை வாய்ந்த அரண்மனை வளாகமும் சீனாவின் அரசு மாளிகை கட்டிடங்களில் ஒன்றுமாகும். 
## 
அரண்மனை அருங்காட்சியகம் இங்கு உள்ளது?
## 
அரண்மனை அருங்காட்சியகம், சீன தலைநகர் பெய்சிங்கின் நடுவில் அமைந்துள்ளது 

***
சிங்கத்திற்கு அரிமா என்ற பெயருண்டு.
##
சிங்கத்தின் வேறு பெயர் கூறுக?
##
அரிமா

***
சிங்கமானது பூனைப் பேரினத்தைச் சேர்ந்தது.
##
சிங்கம் எப்பேறினத்தை சேர்ந்தது?
##
பூனைப் பேரினத்தைச் சேர்ந்தது

***
ஆப்பிரிக்கா கண்டம் உலகின் இரண்டாவது மிகப்பெரிய மற்றும் அதிக மக்கள் தொகை கொண்ட கண்டம் ஆகும்.
##
உலகின் இரண்டாவது பெரிய கண்டம் எது?
##
ஆப்ரிக்கா

***
சாதாரண வழக்கில் நாடு என்ற சொல், தேசம் (பண்பாடு சார்ந்த ஒன்று) மற்றும் அரசு (அரசியல் சார்ந்த ஒன்று) என்னும் இரண்டு கருத்துருக்களையும் குறிக்கப் பயன்படுத்தப்படுகின்றது.
##
நாடு என்ற சொல் எதைக் குறிக்கிறது?
##
தேசம் மற்றும் அரசைக் குறிக்கும்

***
உலகம் (World) எனப்படுவது அனைத்து மனித நாகரிகத்தையும் குறிப்பதாகும்.
##
உலகம் எதைக் குறிக்கும்?
##
உலகம் அனைத்து மனித நாகரிகத்தை குறிக்கும்.

***
வரலாறு என்பது இறந்த காலத்தைப் பற்றி படிக்கின்ற ஒரு பாடப்பிரிவு
##
வரலாறு என்றால் என்ன?
##
இறந்த காலத்தைப் பற்றி படிக்கின்ற ஒரு பாடப்பிரிவு

***
தென்கொரியா என்றழைக்கப்படும் கொரியக்குடியரசு கிழக்கு ஆசியாவிலுள்ள ஒரு நாடாகும்.
##
தென் கொரியா எங்கு உள்ளது?
##
கிழக்கு ஆசியாவில்.

***
ஃஆங்குவல் எழுத்துக்களின் நெடுங்கணக்கில் 51 எழுத்துகள் உள்ளன
##
ஃஆங்குவலில் எத்தனை எழுத்துக்கள் உள்ளன?
##
51 எழுத்துக்கள்.

***
சியோல் தென்கொரிய நாட்டின் தலைநகராகும்.உலகின் மக்கள் தொகை அதிகமுள்ள நகரங்களில் இதுவும் ஒன்றாகும்.
##
தென் கொரியா நாட்டுத் தலைநகர் எது?
##
சியோல்.

***
மட்பாண்டம் என்பது களிமண் வகைகளை கொண்டு மனிதர்கள் தங்கள் தேவவைகளுக்கு பயன் படுத்தும் பொருட்களின் வகை ஆகும்
##
மட்பாண்டம் என்றால் என்ன?
##
களிமண் வகைகளை கொண்டு மனிதர்கள் தங்கள் தேவவைகளுக்கு பயன் படுத்தும் பொருட்களின் வகை ஆகும்.

***
மட்பாண்டம் என்பது பொதுவாக மண்ணால் செய்யப்படும் பொருட்களைக் குறிக்கும்.
##
மட்பாண்டம் என்பது எதைக் குறிக்கிறது?
##
மண்ணால் செய்யப்படும் பொருட்களைக் குறிக்கும்.

***
சூளையில் உள்ள நடராஜர் கோவில் சுமார் 150 ஆண்டுகளுக்கு முன்பு கட்டப்பட்டது.
##
சூலையில் உள்ள நடராஜர் கோவில் எப்போது கட்டப்பட்டது?
##
150 ஆண்டுகளுக்கு முன்பு

***
கனிமம் எனப்படுவது நிலவியல் வழிமுறைகள் மூலம் உருவான இயற்கையான சேர்வை ஆகும்.
##
கனிமம் என்றால் என்ன?
##
நிலவியல் வழிமுறைகள் மூலம் உருவான இயற்கையான சேர்வை ஆகும்.

***
பணி அல்லது தொழில் என்பது ஒருவருடைய வாழ்வாதாரத்திற்கு, வருமானம் ஈட்டக்கூடிய செயல்.
##
வருமானம் ஈட்டக்கூடியச் செயல் எது?
##
வருமானம் ஈட்டக்கூடியச் செயல் தொழில் ஆகும்

***
வட்டி என்பது ஒருவரிடமிருந்து பணம் கடனாகப் பெறுவதற்காகக் கொடுக்கப்படும் வாடகை ஆகும்.
##
வட்டி என்றால் என்ன?
##
ஒருவரிடமிருந்து பணம் கடனாகப் பெறுவதற்காகக் கொடுக்கப்படும் வாடகை ஆகும்.

***
பணத்தின் மதிப்பை அந்த நாணயம் செய்யப்பட்ட தங்கம், வெள்ளி போன்ற மாழையின் (உலோகத்தின்) மதிப்பைக் கொண்டு மதிப்பிடப்பட்டது.
##
பணத்தை எவ்வாறு மதிப்பிடுவர்?
##
அது செய்யப்பட்ட மாழையின் மதிப்பைக் கொண்டு.

***
மண்ணுக்கு மரியாதை என்பது 1995 இல் வெளிவந்த தமிழ் திரைப்படம் ஆகும்.விக்னேஷ் மற்றும் சங்கவி ஆகியோர் முக்கிய கதாபாத்திரங்களில் நடித்திருக்கின்றனர். 
##
மண்ணுக்கு மரியாதை திரைபடத்தில் நடித்த முக்கிய கதாபாத்திரங்கள் யார்?
##
விக்னேஷ் மற்றும் சங்கவி
***
உத்திரம்பட்டு ஊராட்சி , தமிழ்நாட்டின் இராணிப்பேட்டை மாவட்டத்தில் உள்ள காவேரிப்பாக்கம் வட்டாரத்தில் அமைந்துள்ளது.
##
உத்திரம்பட்டு ஊராட்சி எங்கு அமைந்துள்ளது?
##
தமிழ்நாட்டின் இராணிப்பேட்டை மாவட்டத்தில் உள்ள காவேரிப்பாக்கம் வட்டாரத்தில் அமைந்துள்ளது.

***
கொமில்லா மாவட்டம் தெற்காசியாவின் வங்காளதேச நாட்டின் அறுபத்தி நான்கு மாவட்டங்களில் ஒன்றாகும். வங்கதேசத்தின் தென்கிழக்கில் அமைந்த இம்மாவட்டம், டாக்கா நகரத்திலிருந்து நூறு கிலோ மீட்டர் தொலைவில் சிட்டகாங் கோட்டத்தில் உள்ளது.
##
கொமில்லா மாவட்டம் எந்நாட்டில் உள்ளது?
##
தெற்காசியாவின் வங்காளதேச நாட்டின் அறுபத்தி நான்கு மாவட்டங்களில் ஒன்றாகும்.

***
2011-ஆம் ஆண்டின் மக்கள் தொகை கணக்கெடுப்பின் படி, கொமில்லா மாவட்ட மக்கள் தொகை 53,87,288 ஆகும்
##
கொமில்லா மாவட்டம் மக்கள் தொகை எவ்வளவு?
##
53,87,288 ஆகும்

***
வங்காளதேசத்தின் தென்கிழக்கில் அமைந்த கொமில்லா மாவட்டம் 3,085 சதுர கிலோ மீட்டர் பரப்பளவு கொண்டது. 
##
கொமில்லா மாவட்டத்தின் பரப்பளவு பற்றி கூறுக?
##
3,085 சதுர கிலோ மீட்டர் பரப்பளவு கொண்டது.

***
கொமில்லா மாவட்டம் பதினேழு துணை மாவட்டங்களாகப் பிரிக்கப்பட்டுள்ளது. [1] மேலும் இம்மாவட்டம் பதினெட்டு தொகுதிகள் கொண்ட ஒரு நகராட்சி மன்றமும், நூற்றி எண்பது ஊராட்சி ஒன்றியங்களையும், 3624 கிராமங்களையும் கொண்டுள்ளது.  
##
கொமில்லா மாவட்டத்தின் மொத்த தொகுதிகள் எத்தனை?
##
பதினெட்டு தொகுதிகள் கொண்டது.

***
கொமில்லா மாவட்டம் வேளாண்மைப் பொருளாதாரத்தை அடிப்படையாகக் கொண்டுள்ளது. கதர் துணிகள் நெய்வதைக் குடிசைத் தொழிலாக கொண்டுள்ளது.
##

கொமில்லா மாவட்டத்தின் குடிசைத் தொழில் யாது?
##
கதர் துணிகள் நெய்வதைக் குடிசைத் தொழிலாக கொண்டுள்ளது.

***
சாக்சி மாலிக், இந்திய மற்போர் வீராங்கனையாவார். இவர் கிளாஸ்கோவில் நடந்த 2014 பொதுநலவாய விளையாட்டுக்களில் பெண்கள் கட்டற்றவகை மற்போர் 58 கிலோ வகுப்பில் இந்தியா சார்பாக பங்கேற்று வெள்ளிப் பதக்கம் வென்றார்
##
சாக்சி மாலிக் யார்?
##
இந்திய மற்போர் வீராங்கனையாவார்.

***
தமிழ் கம்ப்யூட்டர் தமிழ்நாட்டில் மாதம் இருமுறை வெளிவரும் தமிழ் இதழ். இது கணினியியல், நிரலாக்கம் ஆகிய துறைகளுக்கான சிறப்பு இதழ் ஆகும். 
##
தமிழ் கம்ப்யூட்டர் இதழ் பற்றிக் கூறுக?
##
தமிழ்நாட்டில் மாதம் இருமுறை வெளிவரும் தமிழ் இதழ் ஆகும். இது கணினியியல், நிரலாக்கம் ஆகிய துறைகளுக்கான சிறப்பு இதழ் ஆகும்.

***
இல்டன் சகோதரிகள் அல்லது டெய்சி மற்றும் வயலெட் ஹில்டன் பிறப்பிலேயே உடல்ரீதியில் ஒட்டிப் பிறந்த இரட்டையராவர்.
##
இல்டன் சகோதரிகள் யார்?
##
பிறப்பிலேயே உடல்ரீதியில் ஒட்டிப் பிறந்த இரட்டையராவர்.

***
கல்லாகுளம் வீரமாத்தியம்மன் கோயில் தமிழ்நாட்டில் ஈரோடு மாவட்டம், கல்லாகுளம் என்னும் ஊரில் அமைந்துள்ள அம்மன் கோயிலாகும்.
##
கல்லாகுளம் வீரமாத்தியம்மன் கோயில் எங்கு அமைந்துள்ளது?
##
தமிழ்நாட்டில் ஈரோடு மாவட்டம், கல்லாகுளம் என்னும் ஊரில் அமைந்துள்ளது.

***
படியாக்கத்தேர்வு புனைக்கொள்கை, நோய்த்தொற்றுகளுக்கெதிராக நோயெதிர்ப்புத் தொகுதியானது எவ்வாறு செயலாற்றுகிறது என்பதையும், உடலினுள்புகும் குறிப்பிட்ட எதிர்ப்பிகளை அழிப்பதற்காக எவ்விதம் "பி" மற்றும் "டி" வெள்ளையணுக்கள் தேர்வு செய்யப்படுகின்றன என்பதையும் விளக்கும் மாதிரி வடிவமென்பது பரவலாக ஏற்றுக்கொள்ளப்பட்ட ஒன்றாகும்
##
படியாக்கத்தேர்வு புனைக்கொள்கை பற்றிக் கூறுக?
##
படியாக்கத்தேர்வு புனைக்கொள்கை, நோய்த்தொற்றுகளுக்கெதிராக நோயெதிர்ப்புத் தொகுதியானது எவ்வாறு செயலாற்றுகிறது என்பதையும், உடலினுள்புகும் குறிப்பிட்ட எதிர்ப்பிகளை அழிப்பதற்காக எவ்விதம் "பி" மற்றும் "டி" வெள்ளையணுக்கள் தேர்வு செய்யப்படுகின்றன என்பதையும் விளக்கும் மாதிரி வடிவமென்பது பரவலாக ஏற்றுக்கொள்ளப்பட்ட ஒன்றாகும்.

***
விஷால் சந்திரசேகர் என்பவர் ஒரு இந்திய திரைப்பட இசையமைப்பாளர். இவர் தமிழ், தெலுங்கு படங்களுக்கு இசையமைத்துள்ளார்.
##
விஷால் சந்திரசேகர் யார்?
##
இந்திய திரைப்பட இசையமைப்பாளர். 

***
விசால் தன் ஆறுவயதில் கீபோர்டை வாசிக்கக் கற்றுக்கொள்ள ஆரம்பித்தார். தன் பத்து வயதிலேயே தனியாக நிகழ்ச்சிகளில் வாசிக்க ஆரம்பித்தார்.
##
விசால் எந்த வயதில் கீபோர்டை வாசிக்கக் கற்றுக்கொள்ள ஆரம்பித்தார்?
##
ஆறுவயதில். 

 ***
சில கிறிஸ்தவர்கள் நரகம் என்பது பாவம் செய்பவர்கள் தண்டனை பெறும் துன்பமான இடமாக நம்புகின்றனர். மற்றும் சில கிறிஸ்தவர்கள் கடவுளை காண மோட்சத்திற்கு செல்ல தனது பாவங்களை துன்பங்கள் மூலம் களையும் இடமாக நம்புகின்றனர்.
##
பாவம் செய்பவர்கள் தண்டனை பெறும் துன்பமான இடமாக கிரிதவர்கள் நம்புகின்ற ஊடம் எது?
##
நரகம்

***
ஆச்சார்யா உத்திர பிரதேச மாநிலத்திலுள்ள வாரனாசி மாவட்டத்தின் வாரனாசி பட்டதாரி பிரிவு  தொகுதியில் இருந்து உத்திர பிரதேச சட்டமன்றத்திற்கு 2016 ஆம் ஆண்டு தேர்ந்தெடுக்கப்பட்டார்.
##
ஆச்சார்யா வாரனாசி பட்டதாரி பிரிவு  தொகுதியில் இருந்து உத்திர பிரதேச சட்டமன்றத்திற்கு எந்த ஆண்டால் தேர்ந்தெடுக்கப்பட்டார்?
##
2016
***
உலகப் புலி நாள் என்பது புலி வளம்பேணல் குறித்த விழிப்புணர்வை ஏற்படுத்த ஒவ்வோராண்டும் ஜூலை 29இல் கொண்டாடப்படும் நாளாகும்.
##
உலகப் புலி நாள் ஒவ்வோராண்டும் எந்நாளில் கொண்டாடப்படும்?
##
ஜூலை 29இல்
***
காளிபாளையம் கரியகாளியம்மன் கோயில் தமிழ்நாட்டில் திருப்பூர் மாவட்டம், காளிபாளையம் என்னும் ஊரில் அமைந்துள்ள அம்மன் கோயிலாகும்.
##
காளிபாளையம் கரியகாளியம்மன் கோயில் எந்த மாவட்டத்தில் அமைந்துள்ளது?
##
திருப்பூர் மாவட்டம்
***
பசாயெவ் 2006 சூலை 10ம் நாள் குண்டுவெடிப்பு ஒன்றில் கொல்லப்பட்டார்.
##
பசாயெவ் என்றைக்கு கொல்லப்பட்டார்?
##
2006 சூலை 10ம் நாள்
***
வாராந்தி சுப்பிரமணியசுவாமி கோயில் தமிழ்நாட்டில் புதுக்கோட்டை மாவட்டம், வாராந்தி என்னும் ஊரில் அமைந்துள்ள முருகன் கோயிலாகும்.
##
வாராந்தி சுப்பிரமணியசுவாமி கோயில் எங்கு அமைந்துள்ளது?
##
புதுக்கோட்டை மாவட்டம், வாராந்தி என்னும் ஊரில்.
***
உத்தம புத்திரன் 1958 ஆம் ஆண்டு வெளிவந்த தமிழ்த் திரைப்படமாகும்.
##
உத்தம புத்திரன் எந்த ஆண்டு வெளிவந்தது?
##
1958
***
எல்ஜி எலெக்ட்ரானிக்ஸ் தென் கொரியா நாட்டில் உள்ள சியோல் நகரைத் தலைமையிடமாகக் கொண்ட பன்னாட்டு மின்னணு நிறுவனம்.
##
எல்ஜி எலெக்ட்ரானிக்ஸ் எந்த நாட்டைத் தலைமையிடமாகக் கொண்டது?
##
தென் கொரியா
***
ஜார்ஜ் பிரடெரிக் சாமுவேல் ராபின்சன் 1859ல் ரிப்பன் பிரபு மற்றும் ஏர்ல் டி சாம்பல் என்று அழைக்கப்படுபவர்.
##
1859ல் ரிப்பன் பிரபு என்று அழைக்கப்படுபவர் யார்?
##
ஜார்ஜ் பிரடெரிக் சாமுவேல்
***
சின்னசுவாமி சுப்பிரமணிய பாரதி கவிஞர், எழுத்தாளர், இதழாசிரியர், விடுதலைப் போராட்ட வீரர் மற்றும் சமூக சீர்திருத்தவாதி ஆவார். இவரைப் பாரதியார் என்றும் மகாகவி என்றும் அழைக்கின்றனர்.  
##
மகாகவி என்றும அழைக்கப்படுபவர்‌ யார்?
##
சின்னசுவாமி சுப்பிரமணிய பாரதி.
***
"யாமறிந்த மொழிகளிலே தமிழ்மொழிபோல் இனிதாவ தெங்கும் காணோம்" எனக் கவிபுனைந்தார், பாரதி. 
##
"யாமறிந்த மொழிகளிலே தமிழ்மொழிபோல் இனிதாவ தெங்கும் காணோம்" எனக் கவிபுனைந்தவர் யார்?
##
பாரதி.
 ***
சக்கரவர்த்தினி என்ற மகளிர் மாத இதழிலும் (ஆக. 1905 - ஆக. 1906), இந்தியா என்ற வார இதழில் (மே 1905 - மார்.1906 / செப். 1906, புதுச்சேரி: 10.19.1908 – 17. மே 1910), சூரியோதயம் (1910), கர்மயோகி (திசம்பர் 1909–1910), தர்மம் (பிப்.1910) என்ற இதழ்களிலும் பாலபாரத யங் இண்டியா என்ற ஆங்கில இதழிலும் பாரதி ஆசிரியராகப் பணியாற்றினார்.
##
பாரதி ஆசிரியராகப் பணியாற்றிய மகளிர் மாத இதழ் பெயர்?
##
சக்கரவர்த்தினி
***
பி. சி. ராய் இந்தியாவின் மேற்கு வங்காள மாநிலத்தின் இரண்டாவது முதலமைச்சராக பணியாற்றியவர்.
##
இந்தியாவின் மேற்கு வங்காள மாநிலத்தின் இரண்டாவது முதலமைச்சராக பணியாற்றியவர் யார்?
##
பி. சி. ராய்

***
பி. சி. ராய், 1948ஆம் ஆண்டு முதல் 1962ஆம் ஆண்டில் தமது இறப்பு வரை 14 ஆண்டுகள் காங்கிரசு முதலமைச்சராக தொடர்ந்து இப்பதவியில் இருந்தார்
##
பி. சி. ராய் எத்தனை ஆண்டுகள்  காங்கிரசு முதலமைச்சராக தொடர்ந்து இப்பதவியில் இருந்தார்
##
14 ஆண்டுகள்
***
சூலை 1 தேசிய மருத்துவர் நாள் எனக் கொண்டாடப்படுகிறது. 
##
தேசிய மருத்துவர் நாள் எனக் கொண்டாடப்படு நாள் எது?
##
சூலை 1
***
மேற்கு வங்காளம் என்பது இந்தியாவின் வடகிழக்குப் பகுதியில் அமைந்துள்ள ஒரு மாநிலமாகும்.
##
மேற்கு வங்காளம் என்பது இந்தியாவின் எந்தப் பகுதியில் அமைந்துள்ள ஒரு மாநிலமாகும்?
##
வடகிழக்குப் பகுதியில்
***
பி.டி. ஆர். என்றும் பி.டி. ரணதிவே என்றும் அழைக்கப்பட்ட பாலச்சந்திர திரியம்பக் ரணதிவே(டிசம்பர் 19, 1904 – ஏப்ரல் 6, 1990) ஒரு இந்திய பொதுவுடைமைவாதியும் , தொழிற்சங்கத் தலைவருமாவார் . 
##
பி.டி. ஆர். என்று அழைக்கப்பட்டவர் யார்?
##
பாலச்சந்திர திரியம்பக் ரணதிவே
***
மாரத்தான் போர் கிமு 490 ஆம் ஆண்டில் கிரேக்கம் மீதான பாரசீகர்களின் முற்றுகையின் முதல் கட்டத்தில் இடம் பெற்றது.
##
மாரத்தான் போர் எந்த ஆண்டில் இடம் பெற்றது?
##
கிமு 490 ஆம் ஆண்டில்
***
கிரேக்க நாட்டிற்கு வடக்கே அல்பேனியாவும், மாசிடோனியாவும், பல்கேரியாவும், கிழக்கே துருக்கியும் அமைந்துள்ளது.
##
கிரேக்க நாட்டிற்கு கிழக்கே அமைந்துள்ள நாடு எது ?
##
துருக்கி
***
உரோமைப் பேரரசு பிரித்தானியா, ஸ்பெயின், போர்த்துக்கல், பிரான்சு, இத்தாலி, கிரேக்கம், துருக்கி, ஜேர்மனி மற்றும் எகிப்து என பல்வேறு நாடுகளையும் தனது கைவசம் வைத்திருந்த பண்டைக் காலத்து மாபெரும் பேரரசாகும்.
##
பிரித்தானியா, ஸ்பெயின், போர்த்துக்கல், பிரான்சு, இத்தாலி, கிரேக்கம், துருக்கி, ஜேர்மனி மற்றும் எகிப்து என பல்வேறு நாடுகளையும் தனது கைவசம் வைத்திருந்த பண்டைக் காலத்து மாபெரும் பேரரசு எது?
##
உரோமைப் பேரரசு
***
அகஸ்டஸ் தன்னுடைய வாழ்நாளில் பல்வேறு பெயர்களால் அழைக்கப்பட்டார்.சீஸரின் தத்தெடுப்பிற்குப் பிறகு இவருடைய பெயரினை கயஸ் ஜுலியஸ் சீஸர் ஆச்டேவியஸ் என்றாயிற்று.
##
சீஸரின் தத்தெடுப்பிற்குப் பிறகு அகஸ்டஸ் உடைய பெயர் என்னவாயிற்று?
##
கயஸ் ஜுலியஸ் சீஸர் ஆச்டேவியஸ்
***
ஜூலியஸ் சீசர், உரோமைக் குடியரசின் வீழ்ச்சிக்கும் உரோமைப் பேரரசின் எழுச்சிக்கும் வித்திட்ட நிகழ்வுகளில் முக்கியப் பங்கேற்றார்.
##
உரோமைக் குடியரசின் வீழ்ச்சிக்கும் உரோமைப் பேரரசின் எழுச்சிக்கும் வித்திட்ட நிகழ்வுகளில் முக்கியப் பங்கேற்றவர் யார்?
##
ஜூலியஸ் சீசர்
***
ரைன் ஆறு ஐரோப்பாவின் முக்கியமான ஆறுகளில் ஒன்று. ரைன் , ரைன்: லத்தீன் மொழியில் ரெனஸ் என்றும், ரோமானிய மொழியில் ரீன் என்றும், ஜேர்மன் மொழியில் ரெயின் என்றும், பிரஞ்சு மொழியில் லெ ரைன், டச்சு மொழியில் ரிஜின் என்றும் அழைக்கப்படுகிறது.
##
ரைன் டச்சு மொழியில் எவ்வாறு அழைக்கப்படுகிறது?
##
ரிஜின்
***
நெதர்லாந்தில் ஆடப்படும் அனைத்து விளையாட்டுகளினும் புகழ்பெற்று விளங்குவது கால்பந்து ஆகும்.
##
நெதர்லாந்தில் புகழ்பெற்று விளங்கும் விளையாட்டு எது?
##
கால்பந்து
***
தனிச்செய்யுட் சிந்தாமணி என்னும் நூல் அண்ட சுவாமிக் கவிராயர் என்பவரால் இயற்றப்பட்டது.
##
தனிச்செய்யுட் சிந்தாமணி என்னும் நூல் யாரால் இயற்றப்பட்டது?
##
அண்ட சுவாமிக் கவிராயர்
***
வில்வ பழச்சாறு என்பது வில்வ மரத்தின் பழங்களிலிருந்து பெறப்படும் சாறு. உடல் சூடு தணிக்கும் உண்ணத சாறு இது
##
உடல் சூடு தணிக்கும் உண்ணத சாறு இது?
##
வில்வ பழச்சாறு
***
லிஃப்ட் என்பது 2021ஆம் ஆண்டு தமிழில் வெளிவந்த பரபரப்பூட்டும் திரைப்படமாகும். இது வினீத் வரபிரசாத்தின் அறிமுக இயக்கத்தில் உருவானது. 
##
லிஃப்ட் திரைப்படத்தை இயக்கியவர் யார்?
##
வினீத் வரபிரசாத்தின்
***
நெடுங்காடு சட்டமன்றத் தொகுதி, புதுச்சேரி சட்டமன்றத்துக்கான தொகுதியாகும். இது புதுச்சேரி மக்களவைத் தொகுதிக்கு உட்பட்டது.
##
நெடுங்காடு தொகுதி எந்த சட்டமன்றத்துக்கான தொகுதியாகும்?
##
புதுச்சேரி
***
கொமில்லா மாவட்டம் பதினேழு துணை மாவட்டங்களாகப் பிரிக்கப்பட்டுள்ளது. [1] மேலும் இம்மாவட்டம் பதினெட்டு தொகுதிகள் கொண்ட ஒரு நகராட்சி மன்றமும், நூற்றி எண்பது ஊராட்சி ஒன்றியங்களையும், 3624 கிராமங்களையும் கொண்டுள்ளது.  
##
கொமில்லா மாவட்டத்தின் மொத்த தொகுதிகள் எத்தனை?
##
பதினெட்டு தொகுதிகள்
***
2011-ஆம் ஆண்டின் மக்கள் தொகை கணக்கெடுப்பின் படி, கொமில்லா மாவட்ட மக்கள் தொகை 53,87,288 ஆகும்
##
கொமில்லா மாவட்டம் மக்கள் தொகை எவ்வளவு?
##
53,87,288
***
பணி அல்லது தொழில் என்பது ஒருவருடைய வாழ்வாதாரத்திற்கு, வருமானம் ஈட்டக்கூடிய செயல்.
##
வருமானம் ஈட்டக்கூடியச் செயல் எது?
##
தொழில்
***  
பாக்குநீரிணையை முதன் முதலில் நீந்திக் கடந்தவர் என்ற பெருமை பெற்றவர் முருகுப்பிள்ளை நவரத்தினசாமி.
##  
பாக்குநீரிணையை முதன் முதலில் நீந்திக் கடந்தவர் யார்?
##  
முருகுப்பிள்ளை நவரத்தினசாமி
***
ஃஆங்குவல் எழுத்துக்களின் நெடுங்கணக்கில் 51 எழுத்துகள் உள்ளன
##
ஃஆங்குவலில் எத்தனை எழுத்துக்கள் உள்ளன?
##
51
***
சியோல் தென்கொரிய நாட்டின் தலைநகராகும்.உலகின் மக்கள் தொகை அதிகமுள்ள நகரங்களில் இதுவும் ஒன்றாகும்.
##
தென் கொரியா நாட்டுத் தலைநகர் எது?
##
சியோல்.
***  
பழைய எகிப்து இராச்சியத்தின் வரலாறு கிமு 2686-இல் துவங்கி, கிமு 2181 முடிய விளங்கியது
## 
பழைய எகிப்து இராச்சியத்தின் வரலாறு எப்போது துவங்கியது?
## 
கிமு 2686
***   
புழுப்பாம்பு அனைத்தும் பெண் பாம்புகளாக இருப்பதால், ஆண் துணையின்றி இனப்பெருக்கம் செய்பவையாக உள்ளன.
##
எந்த பாம்பு ஆண் துணையின்றி இனப்பெருக்கம் செய்யும்?
## 
புழுப்பாம்பு.
***
ஆப்பிரிக்கா கண்டம் உலகின் இரண்டாவது மிகப்பெரிய மற்றும் அதிக மக்கள் தொகை கொண்ட கண்டம் ஆகும்.
##
உலகின் இரண்டாவது பெரிய கண்டம் எது?
##
ஆப்ரிக்கா
***
அணிகளைக் கொண்ட ஒரு குழுவில் ஒவ்வோர் அணியும் ஏனைய ஒன்பது அணிகளுடன் ஒரு முறை மோதின.தொடர் முழுவதும் சிறப்பாக ஆடிய வீரர்களைக் கொண்டு தொடரின் அணி உருவாக்கப்படுகிறது. அந்த வகையில் 15 சூலை 2019 நாளன்று கேன் வில்லியம்சன் தலைமையிலான தொடரின் அணியை ஐசிசி அறிவித்தது.
##
யார் தலைமையிலான தொடரின் அணியை ஐசிசி அறிவித்தது?
##
கேன் வில்லியம்சன்
***
புனே விமான நிலையம் புனே நகரத்தில் இருந்து 10 கிலோமீட்டர் தொலைவில் உள்ளது.
##
புனே விமான நிலையம் புனே நகரத்திலிருந்து எவ்வளவு தொலைவில் உள்ளது.
##
10 கிலோமீட்டர்
***
தச்சால் அல்லது மசீக் தச்சால் என்பவன் உலக அழிவின் சமீபத்தில் வெளிப்படும் விசித்திர மனிதன் என்பது இசுலாமியர்களின் நம்பிக்கை ஆகும்
##
உலக அழிவின் சமீபத்தில் வெளிப்படும் விசித்திர மனிதன் என இசுலாமியர்கள் நம்புவது?
##
தச்சால்
***
தினபூமி தமிழ்நாட்டில் வெளியாகும் ஒரு தமிழ் நாளிதழ். 1990 களின் பிற்பகுதியில் வெளியாகத் தொடங்கியது.
##
தினபூமி வெளியாகத் தொடங்கிய ஆண்டு?
##
1990
***
அசிமோ என்பது ஹோண்டா நிறுவனத்தால் உருவாக்கப்பட்ட மனிதனைப்போன்ற தானியங்கிப் பொறி ஆகும். இது 130 சென்டிமீட்டர் உயரமும் 54 கிலோகிராம் எடையும் கொண்டது. 
##
ஹோண்டா நிறுவனத்தால் உருவாக்கப்பட்ட மனிதனைப்போன்ற தானியங்கிப் பொறி பெயர்?
##
அசிமோ
***
ஒய். வி. ராவ் எனப்படும் எறகுடிப்பட்டி வரத ராவ் என்பவர் ஒரு இந்தியத் திரைப்படத் தயாரிப்பாளர், இயக்குநர், மேடை நாடக நடிகர், திரைக்கதை ஆசிரியர், படத்தொகுப்பாளர், நடிகர் ஆவார்.
##
ஒய். வி. ராவ் எனப் அழைக்கப் படும் நபர் யார்?
##
எறகுடிப்பட்டி வரத ராவ்
***
சூரத் அனல் மின் நிலையம் இந்தியாவில் உள்ள குசராத் மாநிலத்தில் உள்ள சூரத் மாவட்டம் நானி நரோலி கிராமத்தில் நிறுவப்பட்டுள்ளது.பழுப்பு நிலக்கரியை அடிப்படை எரிபொருளாகப் பயன்படுத்தும் அனல் மின் நிலையமாக சூரத் அனல் மின் நிலையம் உள்ளது. 
##
சூரத் அனல் மின் நிலையம் எந்த கிராமத்தில் நிறுவப்பட்டுள்ளது
##
நானி நரோலி
***
தமிழ் ஸ்ரார் வானொலி கனடாவில் இருந்து ஒலிபரப்பாகும் 24 மணி நேர தமிழ் வானொலி.இது 2008 ஆம் ஆண்டு ஜூலை 14 இல் தொடங்கப்பட்டது. உலகெங்கும் உள்ள தமிழ் மக்களுக்கு துல்லியமான ஒலிபரப்புடன், சிறப்பான பல நிகழ்ச்சிகளினூடாக சிறந்த பொழுதுபோக்கினை வழங்குகின்றது.  
##
தமிழ் ஸ்ரார் வானொலி ருந்து ஒலிபரப்பாகும்?
##
கனடா
***
ஈரோடு  இராமசாமி பெரியார், திசம்பர் 24, 1973 அன்று தனது 94 ஆம் வயதில் இயற்கை எய்தினார்.
## 
பெரியார் என்றைக்கு இயற்கை எய்தினார்?
##
திசம்பர் 24, 1973
***
முரசாக்கி சிக்கிபு, ஜப்பான் நாட்டில் ஹையன் காலகட்டத்தில் அரசவையில் பணிபுரிந்த பெண் நாவலாசிரியரும் புலவருமாவார். அவர் கி.பி 1000 வாக்கில் எழுதிய "ஜெஞ்சியின் கதை" என்ற புதினத்தால் புகழ்பெற்றவர். இது ஜப்பானின் ஆரம்ப கால புதினங்களில் ஒன்றாகத் திகழ்ந்தது
##
ஜப்பான் நாட்டில் ஹையன் காலகட்டத்தில் அரசவையில் பணிபுரிந்த பெண் நாவலாசிரியர்?
##
முரசாக்கி சிக்கிபு
***
முரசாக்கி சிக்கிபு, ஜப்பான் நாட்டில் ஹையன் காலகட்டத்தில் அரசவையில் பணிபுரிந்த பெண் நாவலாசிரியரும் புலவருமாவார். அவர் கி.பி 1000 வாக்கில் எழுதிய "ஜெஞ்சியின் கதை" என்ற புதினத்தால் புகழ்பெற்றவர். இது ஜப்பானின் ஆரம்ப கால புதினங்களில் ஒன்றாகத் திகழ்ந்தது.
##
கி.பி 1000 வாக்கில் எழுதிய "ஜெஞ்சியின் கதை" என்ற புதினத்தால் புகழ்பெற்றவர் யார்?
##
முரசாக்கி சிக்கிபு
***
கபிலர்மலை பாலசுப்பிரமணியசுவாமி திருக்கோயில் நாமக்கல் மாவட்டம் கபிலர்மலையில் மலை மேல் அமைந்துள்ளது. 
##
கபிலர்மலை பாலசுப்பிரமணியசுவாமி திருக்கோயில் எந்த மாவட்டத்தில் அமைந்துள்ளது?
##
நாமக்கல்
***
சுந்தரர் இயற்றியது திருத்தொண்டர்த் தொகை.நம்பியாண்டார் நம்பியால் எழுதப்பட்ட நூல் திருத்தொண்டர் திருவந்தாதி. இவ்விரு நூல்களையும் அடிப்படையாகக் கொண்டு சேக்கிழாரால் இயற்றப் பட்டதே திருத்தொண்டர் புராணம். இதன் பெருமை காரணமாக இது பெரிய புராணம்என்று அழைக்கப் படுகிறது. இவ்வாசிரியரை மகாவித்துவான் மீனாட்சி சுந்தரனார் "பக்திச் சுவை நனி சொட்டச் சொட்டப் பாடிய கவி வலவ" எனப் பாராட்டியுள்ளார்.
##
சுந்தரர் இயற்றிய நூலின் பெயர் என்ன?
##
திருத்தொண்டர்த் தொகை
***
சுந்தரர் இயற்றியது திருத்தொண்டர்த் தொகை.நம்பியாண்டார் நம்பியால் எழுதப்பட்ட நூல் திருத்தொண்டர் திருவந்தாதி. இவ்விரு நூல்களையும் அடிப்படையாகக் கொண்டு சேக்கிழாரால் இயற்றப் பட்டதே திருத்தொண்டர் புராணம். இதன் பெருமை காரணமாக இது பெரிய புராணம்என்று அழைக்கப் படுகிறது. இவ்வாசிரியரை மகாவித்துவான் மீனாட்சி சுந்தரனார் "பக்திச் சுவை நனி சொட்டச் சொட்டப் பாடிய கவி வலவ" எனப் பாராட்டியுள்ளார்.
##
திருத்தொண்டத் திருவந்தாதியின் ஆசிரியர் யார்?
##
நம்பியாண்டார் நம்பியால்
***
சுந்தரர் இயற்றியது திருத்தொண்டர்த் தொகை.நம்பியாண்டார் நம்பியால் எழுதப்பட்ட நூல் திருத்தொண்டர் திருவந்தாதி. இவ்விரு நூல்களையும் அடிப்படையாகக் கொண்டு சேக்கிழாரால் இயற்றப் பட்டதே திருத்தொண்டர் புராணம். இதன் பெருமை காரணமாக இது பெரிய புராணம்என்று அழைக்கப் படுகிறது. இவ்வாசிரியரை மகாவித்துவான் மீனாட்சி சுந்தரனார் "பக்திச் சுவை நனி சொட்டச் சொட்டப் பாடிய கவி வலவ" எனப் பாராட்டியுள்ளார்.
##
பெரிய புராணத்தை இயற்றியவர் யார்? 
##
சேக்கிழாரால்
***
சுந்தரர் இயற்றியது திருத்தொண்டர்த் தொகை.நம்பியாண்டார் நம்பியால் எழுதப்பட்ட நூல் திருத்தொண்டர் திருவந்தாதி. இவ்விரு நூல்களையும் அடிப்படையாகக் கொண்டு சேக்கிழாரால் இயற்றப் பட்டதே திருத்தொண்டர் புராணம். இதன் பெருமை காரணமாக இது பெரிய புராணம்என்று அழைக்கப் படுகிறது. இவ்வாசிரியரை மகாவித்துவான் மீனாட்சி சுந்தரனார் "பக்திச் சுவை நனி சொட்டச் சொட்டப் பாடிய கவி வலவ" எனப் பாராட்டியுள்ளார்.
##
மீனாட்சி சுந்தரனார் பெரிய புராணத்தின் ஆசிரியரை எவ்வாறு பாராட்டியுள்ளார்? 
##
"பக்திச் சுவை நனி சொட்டச் சொட்டப் பாடிய கவி வலவ"
***
சுந்தரர் இயற்றியது திருத்தொண்டர்த் தொகை.நம்பியாண்டார் நம்பியால் எழுதப்பட்ட நூல் திருத்தொண்டர் திருவந்தாதி. இவ்விரு நூல்களையும் அடிப்படையாகக் கொண்டு சேக்கிழாரால் இயற்றப் பட்டதே திருத்தொண்டர் புராணம். இதன் பெருமை காரணமாக இது பெரிய புராணம்என்று அழைக்கப் படுகிறது. இவ்வாசிரியரை மகாவித்துவான் மீனாட்சி சுந்தரனார் "பக்திச் சுவை நனி சொட்டச் சொட்டப் பாடிய கவி வலவ" எனப் பாராட்டியுள்ளார்.
##
பெரிய புராணத்தின் மறு பெயர் என்ன?
##
திருத்தொண்டர் புராணம்.
"""
```
